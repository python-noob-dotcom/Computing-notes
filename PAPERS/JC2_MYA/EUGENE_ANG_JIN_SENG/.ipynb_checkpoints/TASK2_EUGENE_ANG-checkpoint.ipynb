```python
with open('./MEDIA.txt', 'r') as media:
    data = []
#     Get all the data
    for line in media:
        data += [line.strip().split(',')]
        
data_copy = data
    
def mergesort_notso(array, sortby):
    """
    Implements MergeSort, but for the specified scenario
    """
    if sortby == 'T':
        index_sort = 0
    elif sortby == 'G':
        index_sort = 1
    elif sortby == 'P':
        index_sort = 2
    elif sortby == 'Y':
        index_sort = 3
        
    if len(array) > 1:
        mid = len(array) // 2
        
        left = array[:mid]
        right = array[mid:]
        
        mergesort_notso(left, sortby)
        mergesort_notso(right, sortby)
        
        i = j = k = 0
        
        while i < len(left) and j < len(right):
            if left[i][index_sort] < right[j][index_sort]:
                array[k] = left[i]
                i += 1
                
            else:
                array[k] = right[j]
                j += 1
                
            k += 1
            
        while i < len(left):
            array[k] = left[i]
            i += 1
            k += 1
        
        while j < len(left):
            array[k] = right[j]
            j += 1
            k += 1
            
    return array

def media_sort(data):
    sortby = input('How do you want to sort the media by? Title - T, Genre - G, Producer - P, Year of Production - Y: ')

    while sortby not in ['T', 'G', 'P', 'Y']:
        sortby = input('Please enter the correct coding. Title - T, Genre - G, Producer - P, Year of Production - Y: ')
        
    sort_ed = mergesort_notso(data, sortby.upper())
    print(sort_ed)
    return sort_ed

media_sort(data_copy)
media_sort(data_copy)
```

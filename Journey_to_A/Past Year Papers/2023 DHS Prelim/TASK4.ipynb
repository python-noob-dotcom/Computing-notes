```python
import sqlite3

with sqlite3.connect('election.db') as conn:
	cursor = conn.cursor()

	cursor.execute("BEGIN TRANSACTION;")
	cursor.execute("""CREATE TABLE IF NOT EXISTS "Candidate" (
	"CID"	INTEGER,
	"Name"	TEXT,
	"Gender"	TEXT,
	"Class"	TEXT,
	PRIMARY KEY("CID")
	);""")

	cursor.execute("""CREATE TABLE IF NOT EXISTS "Voter" (
	"VID"	INTEGER,
	"Name"	TEXT,
	"Gender"	TEXT,
	"Class"	TEXT,
	PRIMARY KEY("VID")
	);""")

	cursor.execute("""CREATE TABLE IF NOT EXISTS "Vote" (
	"VID"	INTEGER,
	"CID"	INTEGER,
	FOREIGN KEY("CID") REFERENCES "Candidate"("CID"),
	FOREIGN KEY("VID") REFERENCES "Voter"("VID"),
	PRIMARY KEY("VID","CID")
	);""")
	conn.commit()
```

```python
import csv

with open('CANDIDATES.csv') as f:
    reader = csv.reader(f)

    cands = []

    for line in reader:
        cands += [line]

with open('VOTERS.csv') as f:
    reader = csv.reader(f)

    voters = []

    for line in reader:
        voters += [line]

with open('VOTES.csv') as f:
    reader = csv.reader(f)

    votes = []

    for line in reader:
        votes += [line]


add_cand = 'INSERT INTO Candidate VALUES (?, ?, ?, ?)'
add_voter = 'INSERT INTO Voter VALUES (?, ?, ?, ?)'
add_vote = 'INSERT INTO Vote VALUES (?, ?)'


with sqlite3.connect('election.db') as conn:
    cursor = conn.cursor()
    
    for cand in cands:
        cursor.execute(add_cand, cand)

    for voter in voters:
        cursor.execute(add_voter, voter)
    
    for vote in votes:
        cursor.execute(add_vote, vote)

    conn.commit()
```

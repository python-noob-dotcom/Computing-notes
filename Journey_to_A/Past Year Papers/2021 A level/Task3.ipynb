```python
class Node:
    def __init__(self, data):
        self.data = data
        self.nxt = None

class LinkedList:
    def __init__(self):
        self.head = None
    
    def insert(self, val):
        if self.head == None:
            self.head = Node(val)
        
        else:
            node = Node(val)
            node.nxt = self.head
            self.head = node

    def delete(self, val):
        if self.head == None:
            return None
        
        else:
            if val != self.head.data:
                current = self.head
                previous = current

                while current.data != val:
                    previous = current
                    if current.nxt == None:
                        return None
                    else:
                        current = current.nxt

                previous.nxt = current.nxt
            else:
                self.head = self.head.nxt

            return 1
        
    def search(self, val):
        current = self.head

        while current.data != val:
            if current.nxt == None:
                return False
            else:
                current = current.nxt

        return True
    
    def count(self):
        no = 0

        current = self.head

        while current != None:
            no += 1
            current = current.nxt

        return 
    
    def to_string(self):
        lst = []

        current = self.head

        while current != None:
            lst += [current.data]
            current = current.nxt

        return f'{lst}'

ll = LinkedList()

with open('Task3data.txt') as f:
    data = []
    for line in f:
        data += [int(line.strip())]

for item in data:
    ll.insert(item)

print(ll.to_string())
```

```python
# Task 3.2

class SortedLinkedList(LinkedList):
    def __init__(self):
        super().__init__()

    def insert(self, val):
        node = Node(val)
        if self.head == None:
            self.head = node
            return
        
        if self.head.data > val:
            node.nxt = self.head
            self.head = node
            return
        
        else:
            current = self.head
            previous = current

            while current.data < val and current != None:
                previous = current
                current = current.nxt
                if current == None:
                    previous.nxt = node
                    return
            
            node.nxt = previous.nxt
            previous.nxt = node
            return

sll = SortedLinkedList()

with open('Task3data.txt') as f:
    data = []
    for line in f:
        data += [int(line.strip())]

for item in data:
    sll.insert(item)

print(sll.to_string())
print(sll.search(94))
```

```python
class Queue(LinkedList):
    def __init__(self):
        super().__init__()

    def enqueue(self, integer_val):
        current = self.head
        while current.nxt != None:
            current = current.nxt

        current.nxt = Node(integer_val)

    def dequeue(self):
        if self.head != None:
            to_ret = self.head.data
            self.head = self.head.nxt

            return to_ret
        else:
            return None
        
qu = Queue()

with open('Task3data.txt') as f:
    data = []
    for line in f:
        data += [int(line.strip())]

for item in data:
    qu.insert(item)

print(qu.to_string())

for i in range(5):
    print(qu.dequeue())
```

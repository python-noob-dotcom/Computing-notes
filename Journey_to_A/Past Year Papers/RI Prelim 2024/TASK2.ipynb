```python
def insertion_sort(data, index, ascending):
    for i in range(len(data)):
        j = i - 1
        curr_element = data[i]

        if ascending:
            while curr_element[index] < data[j][index] and j + 1 >= 1:
                temp = data[j + 1]
                data[j + 1] = data[j]
                data[j] = temp
        else:
            while curr_element[index] > data[j][index] and j + 1 >= 1:
                temp = data[j + 1]
                data[j + 1] = data[j]
                data[j] = temp         
    return data

data_list = [(1, 'Alice', 85), (2, 'Bob', 90), (3, 'Charlie', 80)] 
sorted_list = insertion_sort(data_list, 1, ascending=False) 
print(sorted_list) 
```

```python
def task2_2(filename):
    with open(filename) as f:
        data = []
        for row in f:
            data.append(tuple(row.strip().split(',')))
    return data

data = task2_2('./datafiles/CODING_COMPETITION.txt')
sorted = insertion_sort(data, 2, False)
sorted
```

```python
# Task 2.3

def task2_3():
    data = task2_2('./datafiles/CODING_COMPETITION.txt')
    sorted = insertion_sort(data, 3, True)
    countries = {}
    for row in sorted:
        if row[2] not in countries.keys():
            countries[row[2]] = [row]
        else:
            countries[row[2]] += [row]

    for country in countries.keys():
        countries[country] = insertion_sort(countries[country], 3, False)
        with open('SCORE_BY_COUNTRY.txt', 'a') as f:
            for score in countries[country]:
                f.write(f'{score[2]},{score[3]},{score[1]},{score[0]}\n')

task2_3()
```

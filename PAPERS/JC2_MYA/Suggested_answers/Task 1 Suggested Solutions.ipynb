```python
# Task 1.4

class Character:
    
    # Constructor
    def __init__(self, Name, XPos, YPos):
        self.Name = Name  # string
        self.XPos = XPos  # integer
        self.YPos = YPos  # integer

    # Accessor / Getter for self.XPos
    def get_XPos(self):
        return self.XPos
    
    # Accessor / Getter for self.YPos
    def get_YPos(self):
        return self.YPos

    # Mutator / Setter for self.XPos
    def set_XPos(self, val):
        self.XPos = (self.XPos + val) % 10001  # modulo 10001 since there are 10001 positions from 0 to 10000
    
    # Mutator / Setter for self.XPos
    def set_YPos(self, val):
        self.YPos = (self.YPos + val) % 10001

    # Move method for character
    def Move(self, keystroke):
        
        # Up
        if keystroke.upper() == "W":  
            self.set_YPos(10)
        
        # Down
        if keystroke.upper() == "S":
            self.set_YPos(-10)
        
        # Left
        if keystroke.upper() == "A":
            self.set_XPos(-10)
        
        # Right
        if keystroke.upper() == "D":
            self.set_XPos(10)
        
        # Teleport
        if keystroke == ' ':
            
            # Invoke dialog to collect X coordinate for teleportation
            x = input("Enter X coordinate to teleport to: ")          
            
            x_is_int = False  # Flag for terminating validation loop when x is an integer
            
            # Validation loop for X coordinate
            while x_is_int == False:
                
                # Case 1 - x is negative
                if x[0] == '-' and x[1:].isdecimal():
                    x_is_int = True
                    
                # Case 2 - x is positive
                elif x.isdecimal():
                    x_is_int = True
                
                # Case 3 - Invalid x value
                else:
                    x = input("Invalid number! Try again with a whole number: ")
            
            # Invoke dialog to collect Y coordinate
            y = input("Enter Y coordinate to teleport to: ")          
            
            y_is_int = False  # Flag for terminating validation loop when y is an integer
            
            # Validation loop for Y coordinate
            while y_is_int == False:
                
                # Case 1 - y is negative
                if y[0] == '-' and y[1:].isdecimal():
                    y_is_int = True
                    
                # Case 2 - y is positive
                elif y.isdecimal():
                    y_is_int = True
                
                # Case 3 - Invalid y value
                else:
                    y = input("Invalid number! Try again with a whole number: ")
            
            # Set positions
            # Conver x and y to integers
            # Subtract existing position to offset to correct teleporting destination
            self.set_XPos(int(x) - self.get_XPos())
            self.set_YPos(int(y) - self.get_YPos())
```

```python
# Task 1.5

class SpeedyCharacter(Character):  # Inheritance
    
    def __init__(self, Name, XPos, YPos, Multiplier):
        super().__init__(Name, XPos, YPos)  # Inherit constructor of super class
        self.Multiplier = Multiplier  # Introduce additional integer attribute, multiplier
        
    # Polymorphed move method via method overriding
    # Introduct multiplier for up, down, left, right movements
    # Swapped code blocks from collection of X then Y coordinate to Y then X coordinates
    def Move(self, keystroke):
        
        # keystroke is a string
        # converted to upper case to handle both upper and lower case inputs
        
        # Up
        if keystroke.upper() == "W":  
            self.set_YPos(self.Multiplier * 10)
        
        # Down
        if keystroke.upper() == "S":
            self.set_YPos(self.Multiplier * -10)
        
        # Left
        if keystroke.upper() == "A":
            self.set_XPos(self.Multiplier * -10)
        
        # Right
        if keystroke.upper() == "D":
            self.set_XPos(self.Multiplier * 10)
        
        # Teleport
        if keystroke == ' ':

            # Invoke dialog to collect Y coordinate
            y = input("Enter Y coordinate to teleport to: ")          
            
            y_is_int = False  # Flag for terminating validation loop when y is an integer
            
            # Validation loop for Y coordinate
            while y_is_int == False:
                
                # Case 1 - y is negative
                if y[0] == '-' and y[1:].isdecimal():
                    y_is_int = True
                    
                # Case 2 - y is positive
                elif y.isdecimal():
                    y_is_int = True
                
                # Case 3 - Invalid y value
                else:
                    y = input("Invalid number! Try again with a whole number: ")

            
            # Invoke dialog to collect X coordinate for teleportation
            x = input("Enter X coordinate to teleport to: ")          
            
            x_is_int = False  # Flag for terminating validation loop when x is an integer
            
            # Validation loop for X coordinate
            while x_is_int == False:
                
                # Case 1 - x is negative
                if x[0] == '-' and x[1:].isdecimal():
                    x_is_int = True
                    
                # Case 2 - x is positive
                elif x.isdecimal():
                    x_is_int = True
                
                # Case 3 - Invalid x value
                else:
                    x = input("Invalid number! Try again with a whole number: ")
            
            # Set positions
            # Subtract existing position to offset to correct teleporting destination
            self.set_XPos(int(x) - self.get_XPos())
            self.set_YPos(int(y) - self.get_YPos())
```

```python
# Task 1.6

# Instantiate characters
Tom = Character('Tom', 50, 50)
Jerry = SpeedyCharacter('Jerry', 100, 50, 200)

# Get player selection
player = input("Enter T for Tom or J for Jerry: ")
print("")

# Set character for current game
# Conversion to upper case to handle both upper and lower case inputs
if player.upper() == 'T':
    player = Tom
elif player.upper() == 'J':
    player = Jerry
    
print(f'You have selected {player.Name}')
print("")    

# Display starting position
print(f'Starting Position: {player.get_XPos()}, {player.get_YPos()}')
print("")

# Get direction to move
direction_to_move = input("Move the character - W = Up, S = Down, A = Left, D = Right: ")
print("")

# Move character in selected direction
player.Move(direction_to_move)

# Display end position
print(f'Current Position: {player.get_XPos()}, {player.get_YPos()}')
```

```python
import pandas as pd

sales_data = pd.read_csv('./sales_data_sample.csv', encoding='utf-8')
clmn = []
for column in sales_data:
    clmn.append(column)
status_true = pd.DataFrame([], columns= clmn)

for index, row in sales_data.iterrows():
    if row['STATUS'] == 'Shipped':
        status_true.loc[-1] = list(row)
        status_true.index = status_true.index + 1
    
print(status_true.head())
```

```python
import pandas as pd

sales_data = pd.read_csv('./sales_data_sample.csv', encoding='utf-8')
sales_data.columns[12]
```

```python
i = 0

for iter, row in sales_data.iterrows():
    if row['STATUS'] == 'Shipped' and (i % 5 == 0 or i % 3 == 0):
        new_row = list(row)
        new_row[6] = ''
        sales_data.loc[iter] = list(new_row)

    if i % 5 == 0 or i % 3 == 0 or i % 8 == 0:
        new_row = list(row)
        new_row[15] = ''
        sales_data.loc[iter] = list(new_row)

    if i % 3 == 0:
        new_row = list(row)
        new_row[12] = ''
        sales_data.loc[iter] = list(new_row)

    if i % 7 == 0 or i % 2 == 0:
        new_row = list(row)
        new_row[12] = ''
        sales_data.loc[iter] = list(new_row)               

    if i % 7 == 0 or i % 2 == 0:
        new_row = list(row)
        new_row[0] = 'Imagine having no order number'
        sales_data.loc[iter] = list(new_row)


    i += 1

```

```python
print(len(sales_data.index))
```

```python
sales_data.to_csv(path_or_buf= './sales_data_dirty.csv')
```

```python
class LinkedList:
    def __init__(self, data):
        self.data = data
        self.pointer = None

    def insert_rec(self, data):
        def insert_rec_helper(data, pointer):
            if pointer.pointer != None:
                insert_rec_helper(data, pointer.pointer)
            else:
                pointer.pointer = LinkedList(data)
        if self.pointer == None:
            self.pointer = LinkedList(data)
            return
        insert_rec_helper(data, self.pointer)

    def delete_rec(self, data):
        def delete_rec_helper(data, pointer, previous):
            if pointer.data == data:
                previous.pointer = pointer.pointer
                return 'Removed'
            else:
                if pointer != None:
                    delete_rec_helper(data, pointer.pointer, pointer)
                if pointer == None:
                    return -1
                
        if data == self.data:
            self.data = self.pointer.data
            self.pointer = self.pointer.pointer
        else:
            delete_rec_helper(data, self.pointer.pointer, self.pointer)

    def display_rec(self):
        def display_rec_helper(pointer):
            if pointer != None:
                print(pointer.data.getDetails())
                display_rec_helper(pointer.pointer)
            else:
                return
        display_rec_helper(self)

    
```

```python
class StudentLinkedList(LinkedList):
    def __init__(self, data):
        super().__init__(data)
        self.studentCount = 0
        self.courseFreq = {}
    
    def updateCourseFreq(self):
        nxt = self.pointer

        self.courseFreq[self.data.course] = 1

        while nxt != None:
            if nxt.data.course not in self.courseFreq.keys():
                self.courseFreq[nxt.data.course] = 1
            else:
                self.courseFreq[nxt.data.course] += 1
            nxt = nxt.pointer

    def displayCourseFreq(self):
        print(self.courseFreq)

    def insert(self, data):
        super().insert_rec(data)
        self.updateCourseFreq()
        self.studentCount += 1
    
    def delete(self, data):
        super().delete_rec(data)
        self.updateCourseFreq()
        self.studentCount -= 1
    
class StaffLinkedList(LinkedList):
    def __init__(self, data):
        super().__init__(data)
        self.staffCount = 0

    def displayCourseFreq(self):
        print(self.courseFreq)

    def insert(self, data):
        super().insert_rec(data)
        self.staffCount += 1
    
    def delete(self, data):
        super().delete_rec(data)
        self.staffCount -= 1
    
```

```python
class Array:
    def __init__(self, capacity):
        self.capacity = capacity
        self.container = [None for i in range(self.capacity)]

    def add_data(self, new_data, i):
        if i < self.capacity:
            self.container[i] = new_data
            return
        
    def get_data(self, i):
        if i < self.capacity:
            return self.container[i]
        
    def delete_data(self, i):
        if i < self.capacity:
            self.container[i] = None

class CircQueue(Array):
    def __init__(self, capacity):
        super().__init__(capacity)
        self.head = 0
        self.tail = 0

    def enqueue(self, new_data):
        if (self.tail + 1) % self.capacity == self.head:
            return 'Queue is full'
        
        else:
            self.container[self.tail] = new_data
            self.tail = (self.tail + 1) % self.capacity

    def dequeue(self):
        val = self.container[self.head]
        self.container[self.head] = None
        self.head = (self.head + 1) % self.capacity

        if self.head == self.tail:
            self.head = 0
            self.tail = 0

        return val
    
    def get_head(self):
        return self.container[self.head]
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

    def make_left_child(self, lchild):
        self.left = Node(lchild)

    def make_right_child(self, rchild):
        self.right = Node(rchild)

    def get_data(self):
        return self.data
    
    def get_left_child(self):
        return self.left
    
    def get_right_child(self):
        return self.right
    
    
```

```python
Root = Node("the")
Node1 = Node("be")
Node2 = Node("to")
Node3 = Node("of")
Node4 = Node("and")
Node5 = Node("a")
Node6 = Node("in")
Node7 = Node("that")
Node8 = Node("have")
Node9 = Node("it")
Node10 = Node("for")
Node11 = Node("not")
Node12 = Node("on")
Node13 = Node("with")
Node14 = Node("as")

Root.make_left_child(Node1)
Node1.make_left_child(Node2)
Node2.make_left_child(Node3)
Node2.make_right_child(Node4)
Node1.make_right_child(Node5)
Node5.make_left_child(Node6)
Node5.make_right_child(Node7)
Root.make_right_child(Node8)
Node8.make_left_child(Node9)
Node9.make_left_child(Node10)
Node9.make_right_child(Node11)
Node8.make_right_child(Node12)
Node12.make_left_child(Node13)
Node12.make_right_child(Node14)

search = input('Enter a value to search for: ')

Q = CircQueue(10)
Q.enqueue(Root)

while True:
    if Q.tail != Q.head:
        node = Q.dequeue()
        if node.get_data() == search:
            print('Found')
            break
        else:
            if node.get_left_child() != None:
                Q.enqueue(node.get_left_child())
            
            if node.get_right_child() != None:
                Q.enqueue(node.get_right_child())

    else:
        print('Not Found')
        break
```

```python
# Helper class to add all the hours up for each individual name
from datetime import datetime

class organiser:
    def __init__(self, name):
        self.name = name
        self.hours = 0
        self.sheet = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vRjhGr62zj6PxLZD2Wvln7fqB8HGo75tJnFFsIy1gKfNGHiNxL_V4e-dGStzB8vu4xIMvGsKIfuQKmI/pub?gid=1268994803&single=true&output=csv')

    def total_hours(self):

        def convert_hours(time):
            hour, minute, second = time.split(':')

            total_hours = int(hour) + int(minute) / 60 + int(second) / 3600
            
            return total_hours
        
        for index, row in self.sheet.iterrows():
            if row['Name in Full (as in NRIC)'] == self.name and row['Verified']:
                time = row['Hours to be awarded']
                self.hours += convert_hours(time)
        
        return self.hours
    

names = ['Asha Meredith Solomon', 'Ang Jin Seng Eugene']
objs = []

for name in names:
    objs += [organiser(name)]

for obj in objs:
    print(obj.name, obj.total_hours())
```

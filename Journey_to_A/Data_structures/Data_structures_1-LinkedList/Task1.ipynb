```python
class ConnectionNode:
    def __init__(self, data, next= None):
        self.data = data
        self.next = next

class LinkedList:
    def __init__(self):
        self.head = None
    
    def isListEmpty(self):
        return self.head == None
    
    def listLength(self):
        no = 0

        current = self.head

        while current != None:
            current = current.next

            no += 1

        return no
    
    def insertEnd(self, data):

        if self.isListEmpty():
            self.head = ConnectionNode(data)
            return

        new_Node = ConnectionNode(data)

        current = self.head

        while current.next != None:
            current = current.next

        current.next = new_Node

    def insertHead(self, data):

        newNode = ConnectionNode(data)

        if self.isListEmpty():
            self.head = newNode
            return

        newNode.next = self.head.next
        self.head = newNode

    def insertAt(self, data, position):

        i = 0

        if self.isListEmpty():
            self.head = ConnectionNode(data)
            return 

        current = self.head

        while i < position:
            i += 1
            if i >= position:
                break            
            current = current.next

            i += 1

        newNode = ConnectionNode(data)

        newNode.next = current.next

        current.next = newNode

    def deleteAt(self, position):

        i = 0

        current = self.head

        while i < position:
            i += 1
            if i >= position:
                break
            current = current.next

            

        current.next = current.next.next

    def deleteEnd(self):

        current = self.head

        while current.next.next != None:
            current = current.next

        current.next = None

    def printList(self):

        string = ''

        current = self.head

        while current != None:

            if current.next != None:
                string += str(current.data) + ' -> '
            
            else:
                string += str(current.data)

            current = current.next

        print(string)
```

```python
linkedlist = LinkedList()
linkedlist.insertEnd('network')
linkedlist.insertEnd('transmission')
linkedlist.insertEnd('receiver')
linkedlist.printList()
linkedlist.deleteAt(2)
linkedlist.insertHead('wireless')
linkedlist.insertAt('data', 2)
linkedlist.insertAt('router', 2)
linkedlist.deleteEnd()
linkedlist.printList()
```

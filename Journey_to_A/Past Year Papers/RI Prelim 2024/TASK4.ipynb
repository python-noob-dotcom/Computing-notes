```python
import sqlite3

with sqlite3.connect('./datafiles/CLINIC.db') as conn:
    cursor = conn.cursor()
    with open('./datafiles/PATIENT.txt') as f:
        query = "INSERT INTO Patient VALUES (?, ?, ?, ?)"
        for line in f:
            line = line.strip().split(',')
            cursor.execute(query, (line[0], line[1], line[2], line[3]))
    with open('./datafiles/STAFF.txt') as f:
        query = "INSERT INTO Staff VALUES (?, ?, ?, ?)"
        for line in f:
            line = line.strip().split(',')
            cursor.execute(query, (line[0], line[1], line[2], line[3]))
    with open('./datafiles/APPOINTMENT.txt') as f:
        query = "INSERT INTO Appointment VALUES (?, ?, ?, ?, ?)"
        for line in f:
            line = line.strip().split(',')
            cursor.execute(query, (line[0], line[1], line[2], line[3], line[4]))

    conn.commit()
```

```python
with sqlite3.connect('./datafiles/CLINIC.db') as conn:
    cursor = conn.cursor()
    res = cursor.execute("SELECT AppointmentID, Patient.PatientID, Patient.Name,\
                    Staff.StaffID, Staff.Name, AppointmentDate,\
                    Diagnosis FROM Patient INNER JOIN\
                    (Appointment INNER JOIN Staff ON\
                    Appointment.StaffID = Staff.StaffID)\
                    ON Patient.PatientID = Appointment.PatientID").fetchall()
    for record in res:
        print(record)
```

```python
with sqlite3.connect('./datafiles/CLINIC.db') as conn:
    cursor = conn.cursor()
    query = 'SELECT Staff.Name, COUNT(Appointment.AppointmentID) FROM Staff \
        LEFT OUTER JOIN Appointment ON Appointment.StaffID = Staff.StaffID \
            GROUP BY Staff.Name ORDER BY COUNT(Appointment.AppointmentID)\
                DESC'
    
    res = cursor.execute(query).fetchall()

    print(f'{"Name":<15}|{"Number of appointments":<20}')
    print('_'*30)

    for row in res:
        print(f'{row[0]:<15}|{row[1]:<20}')
```

```python
class HashTable:
    def __init__(self, max_size):
        self.max_size = max_size
        self.array = [None for i in range(max_size)]

    def hash(self, val):
        return val % self.max_size
    
    def insert(self, val):
        hash_val = self.hash(val)
        copy = hash_val
        while self.array[hash_val] != None:
           hash_val = (hash_val + 1) % self.max_size
           if hash_val == copy:
               return 'HT is full'  
           
        self.array[hash_val] = val

        return
    
    def display(self):
        print(f'{"Index":<10}{"Item":<15}')
        for i in range(len(self.array)):
            if self.array[i] != None:
                print(f'{i:<10}{self.array[i]:<15}')
            else:
                print(f'{i:<10}{"None":<15}')

    def search(self, val):
        hash_val = self.hash(val)
        copy = hash_val
        while self.array[hash_val] != val:
           hash_val = (hash_val + 1) % self.max_size
           if hash_val == copy:
               return 'Not found'

           else:
                continue
        return hash_val  


ht = HashTable(20)

with open('KEYS2.txt') as f:

    for line in f:
        ht.insert(int(line))

print(ht.search(37))
print(ht.search(77))
print(ht.search(97))
```

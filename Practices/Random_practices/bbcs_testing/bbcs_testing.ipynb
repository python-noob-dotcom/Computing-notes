```python
import pandas as pd

sales_data = pd.read_csv('./sales_data_sample.csv', encoding='utf-8')
clmn = []
for column in sales_data:
    clmn.append(column)
status_true = pd.DataFrame([], columns= clmn)

for index, row in sales_data.iterrows():
    if row['STATUS'] == 'Shipped':
        status_true.loc[-1] = list(row)
        status_true.index = status_true.index + 1
    
print(status_true.head())
```

```python
class Array:
    def __init__(self, n):
        self.data = [None for i in range(n)]
        self.left = [i for i in range(n - 1)]
        self.right = [None for i in range(n - 1)]
        self.right += [None]
        self.left += [None]
        self.free = 0
        self.head = -1

    def get_free(self):
        return self.free
    
    def get_data_index(self, index):
        return self.data[index]
    
    def get_left_index(self, index):
        return self.left[index]
    
    def get_right_index(self, index):
        return self.right[index]
    
    def set_data_index(self, index, value):
        self.data[index] = value

    def set_left_index(self, index, value):
        self.left[index] = value

    def set_right_index(self, index, value):
        self.right[index] = value
```

```python
class BST:
    def __init__(self, n):
        super().__init__(n)

    def add_node(self, data):
        if self.head == -1:
            self.head = self.free
            temp = self.free
            self.free = self.get_left_index(self.free)
            self.set_left_index(temp, None)
            self.set_data_index(temp, data)
            return

        else:
            curr = self.head
            while True:
                if data < self.get_data_index(curr):
                    if self.get_left_index(curr) == None:
                        temp = self.free
                        self.free = self.get_left_index(self.free)
                        self.set_left_index(curr, temp)
                        self.set_data_index(temp, data)
                        self.set_left_index(temp, None)
                        return
                
                    curr = self.get_left_index(curr)
                
                if data > self.get_data_index(curr):
                    if self.get_right_index(curr) == None:
                        temp = self.free
                        self.free = self.get_left_index(self.free)
                        self.set_right_index(curr, temp)
                        self.set_data_index(temp, data)
                        self.set_left_index(temp, None)
                
                    curr = self.get_right_index(curr)

    def search_tree(self, data):
        if self.head == -1:
            print('BST is empty')
            return
        else:
            curr = self.head
            while True:
                if data == self.get_data_index(curr):
                    return curr
                
                elif data < self.get_data_index(curr):
                    if self.get_left_index(curr) == None:
                        print('Data not found')
                        return
                    curr = self.get_left_index(curr)
                
                elif data > self.get_data_index(curr):
                    if self.get_right_index(curr) == None:
                        print('Data not found')
                        return
                    curr = self.get_right_index(curr)

    def in_order_traversal(self, data):
        if self.head == -1:
            print('BST is empty')
            return
        
        def in_order_helper(data, curr):
            datas = []
            if curr == None:
                return datas
            
            else:
                datas += in_order_helper(data, self.get_left_index(curr))
                datas.append(self.get_data_index(curr))
                datas += in_order_helper(data, self.get_right_index(curr))
            
            return datas

        return in_order_helper(data, self.head)    
```

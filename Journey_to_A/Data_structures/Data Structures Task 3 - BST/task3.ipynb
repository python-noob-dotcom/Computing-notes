```python
class Node:
    def __init__(self, data= None, leftPtr= -1, rightPtr= -1):
        self.__data = data
        self.__leftPtr = leftPtr
        self.__rightPtr = rightPtr

    def setData(self, data):
        self.__data = data

    def setLeftPtr(self, idx):
        self.__leftPtr = idx

    def setRightPtr(self, idx):
        self.__rightPtr = idx

    def getData(self):
        return self.__data
    
    def getLeftPtr(self):
        return self.__leftPtr
    
    def getRightPtr(self):
        return self.__rightPtr
    
class Tree:
    def __init__(self, size = 10):
        self.thisTree = [Node() for i in range(size)]
        self.root = -1

    def add(self, newItem):
        if self.root == -1:
            self.root = 0
            current_node = self.thisTree[self.root]
            current_node.setData(newItem)
            return 1
        
        else:
            ptrs = self.root
            prev = ptrs
            idx = 0
            while self.thisTree[idx].getData() != None:
                idx += 1

            self.thisTree[idx].setData(newItem)

            while True:
                if self.thisTree[ptrs].getData() < newItem:
                    if self.thisTree[ptrs].getRightPtr() == -1:
                        self.thisTree[ptrs].setRightPtr(idx)
                        return
                    ptrs = self.thisTree[ptrs].getRightPtr()
                else:
                    if self.thisTree[ptrs].getLeftPtr() == -1:
                        self.thisTree[ptrs].setLeftPtr(idx)
                        return
                    ptrs = self.thisTree[ptrs].getLeftPtr()

    def prnt(self):
        idx = 0
        for node in self.thisTree:
            print(f'(Index: {idx}, Data: {node.getData()}, leftPtr: {node.getLeftPtr()}, rightPts: {node.getRightPtr()})')
            idx += 1

    def postOrderTraversal(self, idx=None):
        if idx is None:
            idx = self.root  # Start from the root if no index is provided

        post_order = []

        # Traverse the left subtree
        left_idx = self.thisTree[idx].getLeftPtr()
        if left_idx != -1:
            post_order += self.postOrderTraversal(left_idx)

        # Traverse the right subtree
        right_idx = self.thisTree[idx].getRightPtr()
        if right_idx != -1:
            post_order += self.postOrderTraversal(right_idx)

        # Visit the current node
        post_order.append(self.thisTree[idx].getData())

        return post_order

    


        
tree = Tree(8)
tree.add('Tiger')
tree.add('Lemur')
tree.add('Bat')
tree.add('Yak')
tree.add('Ostrich')
tree.add('Raccoon')
tree.add('Macaw')
tree.add('Zebra')

tree.prnt()
print(tree.postOrderTraversal(idx= 0))
```

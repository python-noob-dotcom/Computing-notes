```python
class Animal:
    def __init__(self, name, mass, volume, age):
        self.name = name
        self.mass = float(mass)
        self.volume = float(volume)
        self.age = float(age)

    def get_name(self):
        return self.name
    
    def get_mass(self):
        return self.mass

    def get_volume(self):
        return self.volume

    def get_age(self):
        return self.age
    
    def isSick(self):
        return (self.mass / self.volume) < 0.8
    
    def __str__(self):
        return f'Name: {self.name}, Mass: {self.mass}, \
            Volume: {self.volume}, Age: {self.age}'
    
class WarmBloodAnimal(Animal):
    def __init__(self, name, mass, volume, age, temp):
        super().__init__(name, mass, volume, age)
        self.bodyTemp = float(temp)

class ColdBloodAnimal(Animal):
    def __init__(self, name, mass, volume, age, IsPoisonous):
        super().__init__(name, mass, volume, age)
        self.IsPoisonous = IsPoisonous
    
    def getIsPoisonous(self):
        return self.IsPoisonous

class Dog(WarmBloodAnimal):
    def __init__(self, name, mass, volume, age, temp, hasOwner):
        super().__init__(name, mass, volume, age, temp)
        self.hasOwner = hasOwner

    def hasFever(self):
        return self.bodyTemp > 39
    
    def isSick(self):
        return super().isSick() or self.hasFever()
    
    def isYoung(self):
        return self.age < 1.0
    
    def __str__(self):
        return f'Class: Dog, Name: {self.name}, Mass: {self.mass}, \
            Volume: {self.volume}, Age: {self.age}, \
            Body Temp: {self.bodyTemp}, Has Owner: {self.hasOwner}\
            Is Sick: {self.isSick()}, Is Young: {self.isYoung()}'
    
class Goose(WarmBloodAnimal):
    def __init__(self, name, mass, volume, age, temp):
        super().__init__(name, mass, volume, age, temp)

    def hasFever(self):
        return self.bodyTemp > 43.0

    def isSick(self):
        return super().isSick() or self.hasFever()
    
    def isYoung(self):
        return self.age < 0.5

    def __str__(self):
        return f'Class: Goose, Name: {self.name}, Mass: {self.mass}, \
        Volume: {self.volume}, Age: {self.age}, \
        Body Temp: {self.bodyTemp}, \
        Is Sick: {self.isSick()}, Is Young: {self.isYoung()}'
            
class Frog(ColdBloodAnimal):
    def __init__(self, name, mass, volume, age, IsPoisonous):
        super().__init__(name, mass, volume, age, IsPoisonous)
    
    def isYoung(self):
        return self.age < 0.3
    
    def __str__(self):
        return f'Class: Frog, Name: {self.name}, Mass: {self.mass}, \
        Volume: {self.volume}, Age: {self.age}, Is Poisonous: {self.IsPoisonous}, \
        Is Sick: {self.isSick()}, Is Young: {self.isYoung()}'
    
class Snake(ColdBloodAnimal):
    def __init__(self, name, mass, volume, age, IsPoisonous, length):
        super().__init__(name, mass, volume, age, IsPoisonous)
        self.length = float(length)

    def isSick(self):
        return 0.8 * (0.0035 * (self.length * 100) ** 3) > self.mass
    
    def isYoung(self):
        return self.age < 0.5
    
    def __str__(self):
        return f'Class: Snake, Name: {self.name}, Mass: {self.mass}, \
        Volume: {self.volume}, Age: {self.age}, \
        Length: {self.length}, Is Poisonous: {self.IsPoisonous}, \
        Is Sick: {self.isSick()}, Is Young: {self.isYoung()}'
```

```python
# Task 1.4

def task1_4(a_list, animal_type):
    instances = []
    for animal in a_list:
        if animal[0] == animal_type:
            if animal_type == 'Dog':
                instances += [Dog(animal[1], animal[2], animal[3], animal[4], animal[5], animal[6])]

            if animal_type == 'Goose':
                instances += [Goose(animal[1], animal[2], animal[3], animal[4], animal[5])]

            if animal_type == 'Frog':
                instances += [Frog(animal[1], animal[2], animal[3], animal[4], animal[5])]
            
            if animal_type == 'Snake':
                instances += [Snake(animal[1], animal[2], animal[3], animal[4], animal[5], animal[6])]
        
    well_young = 0
    well_adult = 0
    sick_young = 0
    sick_adult = 0

    for animal in instances:
        if animal.isSick():
            if animal.isYoung():
                sick_young += 1
            else:
                sick_adult += 1
        else:
            if animal.isYoung():
                well_young += 1
            else:
                well_adult += 1

    print('-'*20)
    print(f'{animal_type:<10}| Well | Sick |')
    print(f'{"Young":<10}| {well_young:<4} | {sick_young:<4} |')
    print(f'{"Adult":<10}| {well_adult:<4} | {sick_adult:<4} |')
    print('-'*20)

a_list = task1_3() 
task1_4(a_list, 'Dog') 
task1_4(a_list, 'Goose') 
task1_4(a_list, 'Frog') 
task1_4(a_list, 'Snake') 
```

```python
# Task 1.5

def task1_5(a_list, animal_type):
    instances = []
    for animal in a_list:
        if animal[0] == animal_type:
            if animal_type == 'Dog':
                instances += [Dog(animal[1], animal[2], animal[3], animal[4], animal[5], animal[6])]

            if animal_type == 'Goose':
                instances += [Goose(animal[1], animal[2], animal[3], animal[4], animal[5])]

            if animal_type == 'Frog':
                instances += [Frog(animal[1], animal[2], animal[3], animal[4], animal[5])]
            
            if animal_type == 'Snake':
                instances += [Snake(animal[1], animal[2], animal[3], animal[4], animal[5], animal[6])]

    smallest_adult = float('inf')

    for animal in instances:
        if not animal.isYoung():
            if animal.get_volume() < smallest_adult:
                smallest_adult = animal.get_volume()
    
    names = []
    for animal in instances:
        if animal.isYoung():
            if animal.get_volume() > smallest_adult:
                names += [animal.get_name()]

    return names
```

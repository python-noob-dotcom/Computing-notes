```python
class Node:
    def __init__(self, data, nxt_ptr = -1):
        self.data = data
        self.nxt_ptr = nxt_ptr

class SortedLinkedList:
    def __init__(self):
        pass

    def constructor(self):
        self.next_free = 0
        self.start = -1
        self.nodes = [Node(None, i + 1) for i in range(49)]
        self.nodes += [Node(None, -1)]

    def addNode(self, data):
        if self.next_free == -1:
            return
        
        if self.start == -1:
            self.start += 1
            self.nodes[self.next_free].data = data
            self.next_free = self.nodes[self.next_free].nxt_ptr
            print(self.next_free)
            return

        curr = self.start
        prev = curr

        if self.nodes[curr].data > data:
            self.nodes[self.next_free].data = data
            self.nodes[self.next_free].nxt_ptr = self.start
            self.start = self.next_free
            self.nodes[self.next_free].nxt_ptr = -1

        while self.nodes[curr].data < data:
            prev = curr
            curr = self.nodes[curr].nxt_ptr
            if curr == -1:
                self.nodes[self.next_free].data = data
                self.nodes[prev].nxt_ptr = -1
                self.next_free = self.nodes[self.next_free].nxt_ptr
            
        self.nodes[self.next_free].data = data
        next_free = self.nodes[self.next_free].nxt_ptr
        self.nodes[self.next_free].nxt_ptr = curr
        self.nodes[prev].nxt_ptr = self.next_free
        self.next_free = next_free

    def display_content(self):
        curr = self.start
        data = []
        while curr != -1:
            print(curr)
            if self.nodes[curr].data != None:
                data += [self.nodes[curr].data]
            curr = self.nodes[curr].nxt_ptr
        
        print('The items in the list in alphabetical order are:', end= '')
        for i in data:
            print(i, end=', ')
            print('')

        return data

    def findCommon(self, sorted_list):
        data = []
        common = []

        curr = sorted_list.start

        while curr != -1:
            data += [sorted_list.nodes[curr].data]
            curr = sorted_list.nodes[curr].nxt_ptr

        curr = self.start

        while curr != -1:
            if self.nodes[curr].data in data:
                common += [self.nodes[curr].data]

        return common
```

#### General code overview of all the sorts (in a class Sorting())

```python
class Sorting():
    def __init__(self) -> None:
        pass

    def bubble_sort(self, array):

        for i in range(len(array)):

            swapped = False
            
            for j in range(len(array) - i - 1, len(array) - 1):
                if array[j] > array[j + 1]:
                    array[j], array[j + 1] = array[j + 1], array[j]
                    swapped = True

                if not swapped:
                    break
        
        return array
    
    def bubble_sort_version_2(self, array, reverse= False):

        passes = 0
        if not reverse:
            for i in range(len(array) - 1): 

                noSwaps = True
                
                for j in range(len(array) - 1 - i):
                    
                    if array[j] > array[j + 1]:
                        array[j], array[j + 1] = array[j + 1], array[j]
                        noSwaps= False
                    
                passes += 1
                if noSwaps:
                    break
        else:
            for i in range(len(array) - 1): 

                noSwaps= True
                
                for j in range(len(array) - 1 - i):
                    passes += 1
                    if array[j] < array[j + 1]:
                        array[j], array[j + 1] = array[j + 1], array[j]
                        noSwaps = False

                passes += 1
                if noSwaps:
                    break
        
        print(f'Sorted after {passes} passes')
        return array
    
    def bubble_sort_recursive(self, arr, arr_size):

        if arr_size == 1:
            return arr
        
        for i in range(len(arr) - 1):
            if arr[i] > arr[i + 1]:
                arr[i], arr[i + 1] = arr[i + 1], arr[i]

        return self.bubble_sort_recurvise(arr, arr_size - 1)

    def merge_sort(self, array):

        if len(array) > 1:

            mid = len(array) // 2

            left_partition = array[:mid]
            right_partition = array[mid:]

            self.merge_sort(left_partition)
            self.merge_sort(right_partition)

            i = j = k = 0

            while len(left_partition) > i and len(right_partition) > j:
                if left_partition[i] <= right_partition[j]:
                    array[k] = left_partition[i]

                    i += 1 
                else:
                    array[k] = right_partition[j]
                    j += 1 

                k += 1

            while i < len(left_partition):
                array[k] = left_partition[i]
                k += 1
                i += 1
            
            while j < len(right_partition):
                array[k] = right_partition[j]
                k += 1
                j += 1

        return array
    
    def selection_sort(self, array):
        
        for i in range(len(array)):

            minimum = i

            for j in range(i + 1, len(array)):
                if array[j] < array[minimum]:
                    minimum = j

            array[i], array[minimum] = array[minimum], array[i]

        return array
    
    def quick_sort(self, array):
        pass
    
    def insertion_sort(self, array):
        for i in range(1, len(array)):

            curr_element = array[i]

            j = i - 1

            while curr_element < array[j] and j + 1 >= 1:
                array[j + 1] = array[j]
                array[j] = curr_element
                j -= 1

        return array
    

sorts = Sorting()
arr = [5, 4, 6, 5, 4, 5, 8, 9, 2, 1, 7, 5, 10]
print(f'This is sorted with bubble sort {sorts.bubble_sort(arr)}')
print(f'This is sorted with merge sort {sorts.merge_sort(arr)}')
print(f'This is sorted with insertion sort {sorts.insertion_sort(arr)}')
```

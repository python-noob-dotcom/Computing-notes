```python
class Node:
    def __init__(self, data):
        self.data = data
        self.pointer = None
    
class LinkedList:
    def __init__(self):
        self.headpointer = None
    
    def insert_head(self, data):
        if self.headpointer == None:
            self.headpointer = Node(data)
        
        else:
            node = Node(data)

            node.pointer = self.headpointer
            self.headpointer = node
        
        return

    def display(self):
        if self.headpointer == None:
            return 'linked list is empty'

        else:
            curr = self.headpointer

            while curr != None:
                if curr.pointer != None:
                    print(f'{curr.data} --> ', end= '')
                else:
                    print(curr.data)

                curr = curr.pointer
        
    def search(self, name):
        if self.headpointer == None:
            return False

        else:
            curr = self.headpointer

            while curr != None:
                if curr.data[0] == name:
                    return True
                else:
                    curr = curr.pointer
            
            return False

    def remove(self, name):
        if self.headpointer == None:
            return False

        else:
            if self.headpointer.data[0] == name:
                self.headpointer = self.headpointer.pointer
                return True

            curr = self.headpointer
            prev = curr

            while curr != None:
                prev = curr
                if curr.data[0] == name:
                    prev.pointer = curr.pointer
                    return True
                curr = curr.pointer
            
            return False
    
    def reverse_list(self):
        nodes = []
        curr = self.headpointer

        while curr != None:
            nodes += [curr]
            curr = curr.pointer
        
        self.headpointer = nodes[-1]
        curr = self.headpointer

        for i in range(len(nodes) - 2, 0, -1):
            curr.pointer = nodes[i]
            curr = curr.pointer
        
        curr.pointer = nodes[0]
```

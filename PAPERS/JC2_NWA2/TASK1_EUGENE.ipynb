```python
# Task 1.1
class Node:
    def __init__(self, data):
        self.data = data
        self.nxt = None

class linkedlist:
    def __init__(self):
        self.head = None

    def insert(self, data):
        new_node = Node(data)
        if self.head == None:
            self.head = new_node
            return 1
        
        else:
            # Insert in front - duh
            new_node = Node(data)

            new_node.nxt = self.head
            self.head = new_node
            return 1

    def delete(self, data):
        if self.head == None:
            return 'Linked List is empty'
        
        current = self.head

        # Delete specific data. If not in linked list, return none.
        # Iterate through the linked list, search and destroy record.
        while current.nxt != data:
            previous = current
            if current.nxt == None:
                return None
            current = current.nxt
            
        previous.nxt = current.nxt

        return current.data
    
    def search(self, data):
        if self.head == None:
            return 'Linked list is empty'
        
        current = self.head

        # Iterate and search
        while current != None:
            if current.data == data:
                return True
            
            current = current.nxt
            
        return False
    
    def count(self):
        current = self.head

        count = 0

        # Iterate

        while current != None:
            count += 1
            current = current.nxt

        return count

    def to_string(self):
        string = '['

        current = self.head

        # Iterate through the linked list
        while current != None:
            if current.nxt != None:
                string += str(current.data) + ', '
            else:
                string += str(current.data) + ']'
            
            current = current.nxt
        
        return string
```

```python
# Task 1.1
with open('./Task1data.txt', 'r') as f:
    data = f.read().split('\n')

ll = linkedlist()

for char in data:
    ll.insert(int(char))

print(ll.to_string())
```

```python
# Task 1.2

class SortedLinkedList(linkedlist):
    def __init__(self):
        super().__init__()

    def insert(self, data):
        """
        Inserts value into the linked list
        """
        new_node = Node(data)

        if self.head == None:
            self.head = new_node
            return 1
        
        # If the data is less than the head
        elif self.head.data > data:
            new_node.nxt = self.head
            self.head = new_node
            return 1
        
        current = self.head
        # If the data is greater than the head, traverse
        while current.data < data:
            if current.nxt == None:
                current.nxt = new_node
                return 1
            previous = current
            current = current.nxt
        
        # Whack in.
        new_node.nxt = previous.nxt
        previous.nxt = new_node

ll_2 = SortedLinkedList()

for char in data:
    ll_2.insert(int(char))

print(ll_2.to_string())
print(ll_2.search(94))
```

```python
# Task 1.3

class queue(linkedlist):
    def __init__(self):
        super().__init__()

    def enqueue(self, data):
        new_node = Node(data)

        if self.head == None:
            self.head = new_node
            return 1
        
        current = self.head

        while current.nxt != None:
            current = current.nxt

        current.nxt = new_node

        return 1
    
    def dequeue(self):

        if self.head == None:
            return None
        
        dequeued = self.head.data

        self.head = self.head.nxt

        return dequeued
    
q = queue()

for char in data:
    q.enqueue(int(char))

for i in range(5):
    print(q.dequeue())
```

```python
# Task 1.4

class stack(linkedlist):
    def __init__(self):
        super().__init__()

    def push(self, data):
        self.insert(data)

    def pop(self):

        if self.head == None:
            return None
        
        dequeued = self.head.data

        self.head = self.head.nxt

        return dequeued

stac = stack()

for char in data:
    stac.push(int(char))

for i in range(5):
    print(stac.pop())
```

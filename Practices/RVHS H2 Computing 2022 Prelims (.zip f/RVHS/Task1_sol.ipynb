```python
class Node(): 
    # 1
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

class BST(): 
    # 1
    def __init__(self):
        self.root = None

    # 5 [1 + 1 + 1 + 2]
    def insert(self, data):
        new_node = Node(data) #1
        if self.root == None: #1
            self.root = new_node
        else:
            curr = self.root #1
            while curr:
                if curr.data > data: #2
                    if curr.left == None:
                        curr.left = new_node
                        return
                    else:
                        curr = curr.left
                else:
                    if curr.right == None:
                        curr.right = new_node
                        return
                    else:
                        curr = curr.right
    
    def num_leaf(self): #4
        def helper(node):
            if node:
                if not node.left and not node.right:
                    return 1
                else:
                    return helper(node.left) + helper(node.right)
            else:
                return 0
        return helper(self.root)
    
    def preorder(self): #4 [2 + 1 + 1]
        results = ""
        def helper(node):
            nonlocal results
            if node:
                results += str(node.data)
                helper(node.left)
                helper(node.right)
        helper(self.root)
        return results

    def inorder(self): 
        results = ""
        def helper(node):
            nonlocal results
            if node:
                helper(node.left)
                results += str(node.data)
                helper(node.right)
        helper(self.root)
        return results        
    
    def postorder(self):
        results = ""
        def helper(node):
            nonlocal results
            if node:
                helper(node.left)
                helper(node.right)
                results += str(node.data)
        helper(self.root)
        return results

                    
            
def task1_1():
    # code readfile [1]
    # code use bst to store data [2]
    # code to dsplay [1]
    # correct outcome [1]
    f = open("Task1_1.txt", 'r')
    line = f.readline()
    count = 1
    while line:
        bst = BST()
        for letter in line[:-1]:
            bst.insert(letter)
        print(count, bst.num_leaf(), line[:-1], bst.preorder(), bst.inorder(), bst.postorder())
        line = f.readline()
        count += 1
    f.close()
    
task1_1()
```

```python
# Task 1.2

#5
def Task1_2():
    results = []
    f = open("Task1_2.txt", 'r')
    line = f.readline()
    while line:
        no, order, letter_str, order_str, marked = line[:-1].split(',')
        tcno = int(no)
        bst = BST()
        for letter in letter_str:
            bst.insert(letter)
        if order == "preorder":
            if order_str == bst.preorder():
                if marked != "Correct":
                    results.append(tcno)
            else:
                if marked != "Incorrect":
                    results.append(tcno)
        elif order == "inorder":
            if order_str == bst.inorder():
                if marked != "Correct":
                    results.append(tcno)
            else:
                if marked != "Incorrect":
                    results.append(tcno)
        else:
            if order_str == bst.postorder():
                if marked != "Correct":
                    results.append(tcno)
            else:
                if marked != "Incorrect":
                    results.append(tcno)
        line = f.readline()
    f.close()
    return results
    
print(Task1_2())
```

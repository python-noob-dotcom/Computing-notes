```python
class record:
    def __init__(self, name, telephone):
        self.name = name
        self.telephone = telephone

class hashtable:
    def __init__(self):
        self.ds = [None for i in range(500)]
```

```python
with open('HASHEDDATA.txt') as f:
    data = []
    ht = hashtable()
    for line in f:
        line = line.strip().split(',')
        ht.ds[int(line[0])] = record(line[1], line[2])


def DisplayValues(ht):
    print(f'{"Index":<10}|{"PersonName":<20}|{"TelephoneNumber":<30}')
    for i in range(len(ht.ds)):
        print('_'*60)
        print(f'{i:<10}|{ht.ds[i].name:<20}|{ht.ds[i].telephone:<30}')

DisplayValues(ht)
```

```python
def GenerateHash(SearchName):
    HashTotal = 0

    for i in range(len(SearchName)):
        HashTotal += ord(SearchName[i]) * (i + 1)

    return HashTotal % 500

print(GenerateHash('Tait Davinder'))
print(GenerateHash('Anandan Yeo'))
```

```python
def search(ht, searchname):
    hsh = GenerateHash(searchname)

    while ht.ds[hsh].name != searchname :
        hsh += 1
        if ht.ds[hsh].name == '':
            return 'Not Found'
        
    return (hsh, ht.ds[hsh].name, ht.ds[hsh].telephone)

print(search(ht, 'Charlie Love'))
print(search(ht, 'Chin Tan'))
print(search(ht, 'John Barrowman'))
```

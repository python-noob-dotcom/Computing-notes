```python
class QueueItem:
    def __init__(self, id= None, quantity= None):
        self.id = id
        self.quantity = quantity

    def getID(self):
        return self.id
    
    def getQuantity(self):
        return self.quantity
    
```

```python
class Queue:
    def __init__(self):
        self.queueList = [QueueItem() for i in range(4)]
        self.headpointer = 0
        self.tailpointer = 0
        self.numberItems = 0

    def enqueue(self, item):
        print(self.tailpointer)
        if self.tailpointer == 0:
            self.queueList[self.tailpointer] = item
            self.tailpointer = (self.tailpointer + 1) % 4
            self.numberItems += 1
            return True
        
        nxt = self.tailpointer
        if nxt % 4 == 0:
            return False
        
        self.queueList[nxt] = item
        self.tailpointer = nxt + 1
        self.numberItems += 1

        return True
    
    def printQueue(self):
        for item in self.queueList:
            print(f'{item.getID()} -->')

queue = Queue()

for i in range(5):
    result = queue.enqueue(QueueItem(input('Enter an ID: '), input('Enter a quantity')))
    print('Successful' if result is True else 'Failure, critical')

queue.printQueue()
```

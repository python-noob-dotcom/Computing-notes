```python
import csv

with open('marathon.csv') as f:
    reader = csv.reader(f)
    header = next(reader)

    data = []
    dnfs = 0
    finish = 0
    total = 0

    for line in reader:
        data += [line]
        
        if line[2] == 'DNF':
            dnfs += 1
        else:
            finish += 1

        total += 1
        
print(f'Number of DNF: {dnfs}')
print(f'Total number of athletes:{total}')
print(f'Percentage of finishers: {finish/total*100:.2f}%')
```

```python
def insertionsort(lst):
    global key
    for i in range(1, len(lst)):
        
        curr = lst[i]

        j = i - 1

        while j > -1 and lst[j][key] > lst[j + 1][key]:
            lst[j + 1], lst[j] = lst[j], lst[j + 1]
            j -= 1
        
        lst[j + 1] = curr

    return lst
```

```python
key = 2

sorted_data = insertionsort(data)

print(f'{"Rank":<10}{"Country":<15}{"Name":<15}{"Timing":<10}')
for i in range(len(sorted_data)):
    line = sorted_data[i]
    print(f'{i:<10}{line[1]:<15}{line[0]:<15}{line[2]:<10}')
```

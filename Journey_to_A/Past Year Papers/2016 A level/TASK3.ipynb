```python
class Node:
    def __init__(self):
        self.data = None 
        self.left = None
        self.right = None

    def constructor(self):
        pass

    def setData(self, s):
        self.data = s

    def setLeftPtr(self, x):
        self.left = x

    def setRightPtr(self, x):
        self.right = x

    def getData(self):
        return self.data
    
    def getLeftPtr(self):
        return self.left
    
    def getRightPtr(self):
        return self.right
    

class Tree:
    def __init__(self):
        self.tree = [Node() for i in range(10)]
        self.root = -1

    def add(self, newItem):
        if self.root == -1:
            self.root += 1

            self.tree[self.root].setData(newItem)

        else:
            curr = self.root
            # find next open spot
            j = curr

            while j < len(self.tree) and self.tree[j].getData() != None:
                j += 1
                if j == len(self.tree):
                    return 'Full'

            while True:
                if newItem < self.tree[curr].getData():
                    if self.tree[curr].getLeftPtr() == None:
                        self.tree[j].setData(newItem)
                        self.tree[curr].setLeftPtr(j)
                        break
                    else:
                        curr = self.tree[curr].getLeftPtr()
                        
```

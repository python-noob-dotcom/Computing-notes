```python
def read_from_file(filename):
    data = []
    if filename[-4:] == '.txt':
        with open(filename, 'r') as f:
            for row in f:
                data.append(int(row.strip()))
        return data

    else:
        print('Please input the filename of a txt file!')
        return 'Exit with Code 0'
    
```

```python
class sorting():
    def __init__(self) -> None:
        pass
    def bubble_sort(self, array):
        for i in range(len(array) - 1):
            noSwaps = True

            for j in range(len(array) - i - 1):
                if array[j] < array[j + 1]:
                    array[j], array[j + 1] = array[j + 1], array[j]

                    noSwaps = False

            if noSwaps:
                break
        
        return array

    def insertion_sort(self, array):
        for i in range(len(array) - 1):
            curr_element = array[i]

            j = i - 1

            while curr_element > array[j] and j + 1 >= 1: # we use j here as i is the outer loop, and in this inner loop, it should be kept constant
                array[j + 1] = array[j]
                array[j] = curr_element
                j -= 1
        
        return array
    
    def merge_sort(self, array):
        if len(array) > 1:
            mid = len(array) // 2
            left = array[:mid]
            right = array[mid:]

            self.merge_sort(left)
            self.merge_sort(right)

            i = j = k = 0

            while i < len(left) and j < len(right):
                if left[i] > right[j]:
                    array[k] = left[i]
                    i += 1
                else:
                    array[k] = right[j]
                    j += 1
                k += 1

            while i < len(left):
                array[k] = left[i]
                k += 1
                i += 1
            
            while j < len(right):
                array[k] = right[j]
                j += 1
                k += 1

        return array
    
sorts = sorting()

sorts.insertion_sort(data)
```

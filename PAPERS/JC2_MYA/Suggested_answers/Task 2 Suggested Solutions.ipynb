```python
# Task 2.1

# List for storing contents of MEDIA.txt
records = []

# Read contents into list records using 1 sub-list per line
with open('MEDIA.txt', 'r') as f:
    for line in f:
        record = line.strip().split(',')
        records.append(record)

def mergeSort(ds, field):

    # Determine index in sublist for the field to sort by
    if field.upper() == "T":
        index = 0
    elif field.upper() == "G":
        index = 1
    elif field.upper() == "P":
        index = 2
    elif field.upper() == "Y":
        index = 3
      
    # If list or sublist contains more than 1 element
    # Proceed to split up the list or sublist to two smaller halves
    if len(ds) > 1:
        
        # Find middle of the list
        mid = len(ds)//2
        
        # Dividing the list to left and right sub-lists
        left = ds[:mid]
        right = ds[mid:]
        
        # Call to recursively break the sub-lists
        # Before proceeding to sorting elements into the correct position        
        mergeSort(left, field)
        mergeSort(right, field)

        # Counters for maintaining index of left (i) and right sub-lists (j) 
        # As well as the main list (k)
        i = j = k = 0

        # When the list has been completely splitted to individual elements
        # Compare current smallest element in left sub-list and right sub-list
        # Place it in the correct position of the main list
        while i < len(left) and j < len(right):
            if left[i][index] < right[j][index]:
                ds[k] = left[i]
                i = i + 1
            else:
                ds[k] = right[j]
                j = j + 1
            k = k + 1

        # When run out of elements in either the left or right sub-lists
        # Place remaining elements in the other list to the main list
        while i < len(left):
            ds[k] = left[i]
            i = i + 1
            k = k + 1

        while j < len(right):
            ds[k] = right[j]
            j = j + 1
            k = k + 1
        
    return ds 
    
print('--------------------------------------------------------------------------------------------------------------')
print('Sort by Title')
print('--------------------------------------------------------------------------------------------------------------')

field = input('''Enter the criteria to sort by
                Title - T
                Genre - G
                Producer - P
                Year of Production - Y
                ''')

ds_sorted = mergeSort(records, field)
for record in ds_sorted:
    print(record)
    
print('')
    
print('--------------------------------------------------------------------------------------------------------------')
print('Sort by Year')
print('--------------------------------------------------------------------------------------------------------------')

field = input('''Enter the criteria to sort by
                Title - T
                Genre - G
                Producer - P
                Year of Production - Y
                ''')

ds_sorted = mergeSort(records, field)
for record in ds_sorted:
    print(record)
```

```python
# Task 4.1

import sqlite3

with sqlite3.connect(r'C:\Users\eugen\Desktop\Computing_notes\Computing-notes\Journey_to_A\Web_application\Web Applications Practice 2\MerlionThemePark.db') as conn:
    cursor = conn.cursor()

    cursor.execute("BEGIN TRANSACTION")
    cursor.execute('CREATE TABLE "Ticket" (\
	"Date"	TEXT UNIQUE,\
	"dayOfWeek"	TEXT,\
	"unitPrice"	INTEGER,\
	"totQuan"	INTEGER,\
	"availQuan"	INTEGER,\
	PRIMARY KEY("Date")\
    );')

    cursor.execute('CREATE TABLE "Sale" (\
	"saleID"	INTEGER UNIQUE,\
	"tDate"	INTEGER,\
	"quan"	INTEGER,\
	"totalPrice"	INTEGER,\
	FOREIGN KEY("tDate") REFERENCES "Ticket"("Date"),\
	PRIMARY KEY("saleID" AUTOINCREMENT)\
    );')

    conn.commit()
```

```python
# Task 4.2

import sqlite3

with open(r'C:\Users\eugen\Desktop\Computing_notes\Computing-notes\Journey_to_A\Web_application\Web Applications Practice 2\Task4_2_2.txt', 'r') as f:
    sales = []
    for line in f:
        sales += [line.strip().split(',')]

with open(r'C:\Users\eugen\Desktop\Computing_notes\Computing-notes\Journey_to_A\Web_application\Web Applications Practice 2\Task4_2_1.txt', 'r') as f:
    tickets = []
    for line in f:
        tickets += [line.strip().split(',')]

ticket_query = "INSERT INTO Ticket VALUES (?, ?, ?, ?, ?)"
sales_query = "INSERT INTO Sale(saleID, tDate, quan, totalPrice) VALUES (?, ?, ?, ?)"

with sqlite3.connect(r'C:\Users\eugen\Desktop\Computing_notes\Computing-notes\Journey_to_A\Web_application\Web Applications Practice 2\MerlionThemePark.db') as conn:
    cursor = conn.cursor()

    for ticket in tickets:
        cursor.execute(ticket_query, (ticket[0], ticket[1], int(ticket[2]), int(ticket[3]), int(ticket[4])))
    for sale in sales:
        cursor.execute(sales_query, (int(sale[0]), sale[1], int(sale[2]), int(sale[3])))

    conn.commit()
```

```python
# Task 4.3

import sqlite3

month = input('Please input a month to check ticket information for: ')

while not month.isdigit() and month < '13' and month > '0':
    month = input('Please enter a valid month in numbers. ')

month = int(month)
query = "SELECT * FROM Ticket WHERE Ticket.Date LIKE ?"

with sqlite3.connect(r'C:\Users\eugen\Desktop\Computing_notes\Computing-notes\Journey_to_A\Web_application\Web Applications Practice 2\MerlionThemePark.db') as conn:
    cursor = conn.cursor()

    res = cursor.execute(query, (f'%{month}%',)).fetchall()

    print(f'{"Date":<15}|{"Day of Week":<15}|{"unitPrice":<10}|{"totQuan":<10}\
          |{"availQuan":<10}')
    print('_'*100)
    
    for row in res:
        print(f'{row[0]:<15}|{row[1]:<15}|{row[2]:<10}|{row[3]:<10}\
          |{row[4]:<10}')
        print('_'*100)
```

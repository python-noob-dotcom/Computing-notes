## <font color = blue>  Task 1 - Monetary Statistics

Write a program that:

- asks the user to input $n$ integer values of money in dollars
- store the input values in a list and output the list
- display the lowest, highest, sum and mean of the $n$ values entered in dollars.

Given below is a set of sample input and output values.

Quantity of values of money to be entered: `4`<br>
`$100`<br>
`$200`<br>
`$300`<br>
`$400`<br>

The list of values of money entered: `[$100, $200, $300, $400]`<br>

Minimum value is `$100`, maximum value is `$400`, the sum is `$1000` and the mean is `$250.0`.

```python
# Your code here
def myprogram():
    def take_user_input():
        userin1 = input('Quantity of values of money to be entered: ')
        values = []
        for i in range(0, int(userin1)):

            values.append(float(input(' ').strip('$')))
        
        return values
    values = take_user_input()
    totalval = 0
    counter = 0
    for i in values:
        totalval += i
        counter += 1
    mean = totalval / counter
    return f'max ${max(values)}', f'min ${min(values)}', f'sum ${sum(values)}', f'mean ${mean}'

myprogram()
```

Use the built-in `.count()` method to count the occurrences of each integer.

```python
# Your code here
from random import randint

def myprogram2():
    def generate_numbers():
        randomint = []
        for i in range(501):
            randomint.append(randint(1, 100))
        return randomint
    random_integers = generate_numbers()
    print(f'|{"Integer":<15}|{"Occurences":<15}|')
    print(f'|{"---------":<15}|{"---------":<15}|')
    for i in range(1, 101):
        counter = random_integers.count(i)
        print(f'|{i:<15}|{counter:<15}|')
        print('_'*33)
        

myprogram2()
```

## <font color = blue> Task 4 - Pig Latin
    
**The list data structure must be used in your program.**

Pig Latin is a language constructed by transforming English words. 

While the origins of the language are unknown, it is mentioned in at least two documents from the 19th century, suggesting that it has existed for more than 100 years. 

The following rules are used to translate English into Pig Latin: 

- If the word begins with a consonant (including y), then all letters at the beginning of the word, up to the first vowel (excluding y), are removed and then added to the end of the word, followed by ay. 
    - For example, computer becomes omputercay and think becomes inkthay. 
    
    
- If the word begins with a vowel (not including y), then way is added to the end of the word. 
    - For example, algorithm becomes algorithmway and office becomes officeway. 
    
Write a program that reads a line of text from the user and translates the line into Pig Latin, displaying the result.

Your program needs to correctly handle uppercase letters and punctuation marks such as commas, periods, question marks and exclamation marks. 

- If an English word begins with an uppercase letter then its Pig Latin representation should also begin with an uppercase letter and the uppercase letter moved to the end of the word should be changed to lowercase.
    - For example, Computer should become Omputercay. 
    
    
- If a word ends in a punctuation mark then the punctuation mark should remain at the end of the word after the transformation has been performed. 
    - For example, Science! should become Iencescay!.

```python
# Your code here
import re

def PigLatin():
    def list_construction():
        consonants = ['B', 'C', 'D', 'F', 'G', 'H', 'J', 'K', 'L', 'M', 'N', 'P', 'Q', 'R', 'S', 'T', 'V', 'W', 'X', 'Y', 'Z']
        consonants_1 = []
        for i in consonants:
            consonants_1.append(i.lower())
        consonants += consonants_1
        vowels = ['A', 'E', 'I','O', 'U']
        vowels_1 = []
        for i in vowels:
            vowels_1.append(i.lower())
        vowels += vowels_1
        punctuations = '''!()-[]{};:'"\,<>./?@#$%^&*_~'''

        return consonants, vowels, punctuations
    
    consonants, vowels, punctuations = list_construction()
    userin = input('Input a word to be translated into PigLatin')
    userin = list(userin)
    punctuation = []
    if userin[0] in consonants:
        for char in userin:
            if char in punctuations:
                userin.remove(char)
                punctuation.append(char)
        for char in userin:
            if char in vowels:
                start_shift = userin.index(char)
                break
        
        string_slice_1 = userin[:start_shift]
        to_shift = userin[start_shift:]
        final_text = to_shift + string_slice_1 + list('ay')
        for i in punctuation:
            final_text += i
        
    elif userin[0] in vowels:
        for char in userin:
            if char in punctuations:
                userin.remove(char)
                punctuation.append(char)
        final_text = userin + list('way') + punctuation
        for i in punctuation:
            final_text += i
    
    a = ''
    for char in final_text:
        a += char
    
    return a

PigLatin()
```

```python
import sqlite3

with sqlite3.connect('SUPERMARKET.db') as conn:
    cursor = conn.cursor()

    cursor.execute('BEGIN TRANSACTION')
    
    cursor.execute("""    CREATE TABLE "Item" (
        "ItemID"	INTEGER,
        "Name"	TEXT,
        "Price"	TEXT,
        PRIMARY KEY("ItemID")
    );""")

    cursor.execute("""CREATE TABLE "Customer" (
        "CustomerNumber"	INTEGER,
        "Name"	TEXT,
        PRIMARY KEY("CustomerNumber")
    );""")
    
    cursor.execute("""CREATE TABLE "Purchase" (
        "PurchaseID"	INTEGER,
        "CustomerNumber"	INTEGER,
        "ItemID"	INTEGER,
        "Quantity"	INTEGER,
        "DateTime"	TEXT,
        FOREIGN KEY("CustomerNumber") REFERENCES "Customer"("CustomerNumber"),
        FOREIGN KEY("ItemID") REFERENCES "Item"("ItemID"),
        PRIMARY KEY("PurchaseID")
    );""")

    conn.commit()
```

```python
with open('ITEM.txt') as f:
    items = []

    for line in f:
        items += [line.strip().split(',')]

with open('CUSTOMER.txt') as f:
    cust = []

    for line in f:
        cust += [line.strip().split(',')]

with open('PURCHASE.txt') as f:
    pur = []

    for line in f:
        pur += [line.strip().split(',')]

with sqlite3.connect('SUPERMARKET.db') as conn:
    cursor = conn.cursor()

    insert_item = 'INSERT INTO Item VALUES (?, ?, ?)'
    insert_cust = 'INSERT INTO Customer VALUES (?, ?)'
    insert_pur = 'INSERT INTO Purchase VALUES (?, ?, ?, ?, ?)'

    for item in items:
        cursor.execute(insert_item, item)
    
    for cus in cust:
        cursor.execute(insert_cust, cus)

    for pu in pur:
        cursor.execute(insert_pur, pu)

    conn.commit()
```

```python
id = input('Enter the ID of cust: ')
date = input('Enter date that you want to search for: ')


with sqlite3.connect('SUPERMARKET.db') as conn:
    cursor = conn.cursor()

    data = cursor.execute("""
    SELECT Item.Name, Quantity FROM (Item INNER JOIN Purchase
    ON Item.ItemID = Purchase.ItemID) AS A INNER JOIN Customer ON
    A.CustomerNumber = Customer.CustomerNumber WHERE A.CustomerNumber = ?
    AND DateTime >= ?

""", (int(id), date)).fetchall()
    
    for row in data:
        print(f'{row[0]:<10}|{row[1]:<10}')

```

```python
class Node:
    def __init__(self, data, nxt_ptr = -1):
        self.data = data
        self.nxt_ptr = nxt_ptr

class SortedLinkedList:
    def __init__(self):
        pass

    def constructor(self):
        self.next_free = 0
        self.start = -1
        self.nodes = [Node(None, i + 1) for i in range(49)]
        self.nodes += [Node(None, -1)]

    def addNode(self, data):
        if self.next_free == -1:
            return
        
        if self.start == -1:
            self.start += 1
            self.nodes[self.next_free].data = data
            temp = self.nodes[self.next_free].nxt_ptr
            self.nodes[self.next_free].nxt_ptr = -1
            self.next_free = temp
            return

        curr = self.start
        prev = curr

        if self.nodes[curr].data > data:
            self.nodes[self.next_free].data = data
            temp = self.nodes[self.next_free].nxt_ptr
            self.nodes[self.next_free].nxt_ptr = self.start
            self.start = self.next_free
            self.next_free = temp
            return

        while self.nodes[curr].data < data:
            prev = curr
            curr = self.nodes[curr].nxt_ptr
            if curr == -1:
                self.nodes[self.next_free].data = data
                self.nodes[prev].nxt_ptr = -1
                self.next_free = self.nodes[self.next_free].nxt_ptr
            
        self.nodes[self.next_free].data = data
        next_free = self.nodes[self.next_free].nxt_ptr
        self.nodes[self.next_free].nxt_ptr = curr
        self.nodes[prev].nxt_ptr = self.next_free
        self.next_free = next_free
        return

    def display_content(self):
        curr = self.start
        data = []
        while curr != -1:
            print(curr)
            if self.nodes[curr].data != None:
                data += [self.nodes[curr].data]
            curr = self.nodes[curr].nxt_ptr
        
        print('The items in the list in alphabetical order are:')
        for i in data:
            print(i, end=', ')
            print('')

        return data

    def findCommon(self, sorted_list):
        data = []
        common = []

        curr = sorted_list.start

        while curr != -1:
            data += [sorted_list.nodes[curr].data]
            curr = sorted_list.nodes[curr].nxt_ptr

        curr = self.start

        while curr != -1:
            if self.nodes[curr].data in data:
                common += [self.nodes[curr].data]

        return common
```

```python
def task2_2():
    ll = SortedLinkedList()
    ll.constructor()
    ll.addNode('Orange')
    ll.addNode('Banana')
    ll.addNode('Grapes')
    ll.addNode('Apple')
    ll.addNode('Lemon')
    ll.addNode('Eggplant')

    ll.display_content()

task2_2()
```

```python
def task2_3(filename):
    with open(filename, 'r') as f:
        data = []
        for line in f:
            data = line.strip().split(',')

    return data

itemlst1 = task2_3('Task2a.txt') 
print(itemlst1) 
itemlst2 = task2_3('Task2b.txt') 
print(itemlst2)
```

```python
def task2_4():
    itemlst1 = task2_3('Task2a.txt') 
    itemlst2 = task2_3('Task2b.txt')
    sl1 = SortedLinkedList()
    sl1.constructor()
    sl2 = SortedLinkedList()
    sl2.constructor()

    for item in itemlst1:
        sl1.addNode(item)
    
    for item in itemlst2:
        sl2.addNode(item)

    common = sl1.findCommon(sl2)

    return common
```

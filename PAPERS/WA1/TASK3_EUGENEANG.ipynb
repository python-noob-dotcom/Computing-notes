```python
## Task 3.2

def check(data, mode, startyear, endyear):
    """
    Based on the parameters (mode, startyear, endyear) , outputs Bar graphs of the ridership of the mode of transportation
    """
    ridership = []
    ttl = 0
    
#     Print header row
    print(f"{mode:<15}{'Ridership':<15}\n")
    
    for line in data:
        if int(line[0]) >= int(startyear) and int(line[0]) <= int(endyear) and line[1] == mode:
            ridership.append([line[0], line[2]])
#         Generate bar graph
            bar = int((int(line[2]) / 100000)) * '*'
            print(f'{line[0]:<15}{bar:<15}')
#             Get total ridership
            ttl += int(line[2])

    print(f"Average ridership per year is {int(round(ttl/len(ridership), 0))}\n{max(ridership, key= lambda x: int(x[1]))[0]} has the highest ridership of {max(ridership, key= lambda x: int(x[1]))[1]}\n{min(ridership, key= lambda x: int(x[1]))[0]} has the lowest ridership of {min(ridership, key= lambda x: int(x[1]))[1]}")

    

#     Main code
data = []
with open('ridership-Copy1.txt', 'r') as f:
    for line in f:
        data.append(line.strip().split(','))

data.pop(0)
modes = {'M': 'MRT', 'L': 'LRT', 'B': 'Bus', 'T': 'Taxi', 'Q': 'quit'}

while True:
    mode = modes[input("\nSelect M - MRT, L- LRT, B- Bus, T- Taxi or 'Q' to quit >>")]
    if mode != 'quit' and mode in modes.values():
        startyear = input('Key in the start year (2001 to 2016) >> ')
        endyear = input('Key in the end year (2001 to 2016) >> ')
        check(data, mode, startyear, endyear)
        
    else:
        raise Exception('User Quit the program or User entered invalid mode')
```

```python
import datetime

class Member:
    def __init__(self, memberID, first_name, surname, contact_number, memberType):
        self.memberID = memberID
        self.first_name = first_name
        self.surname = surname
        self.contact_number = contact_number
        self.last_visit = datetime.datetime.now()
        self.memberType = memberType

    def get_memberID(self):
        return self.memberID
    
    def set_memberID(self, memberID):
        self.memberID = memberID

    def get_first_name(self):
        return self.first_name
    
    def set_first_name(self, first_name):
        self.first_name = first_name

    def get_surname(self):
        return self.surname
    
    def set_surname(self, surname):
        self.surname = surname

    def get_contact_number(self):
        return self.contact_number

    def set_contact_number(self, contact_number):
        self.contact_number = contact_number
    
    def get_last_visit(self):
        return self.last_visit

    def set_last_visit(self, last_visit):
        self.last_visit = last_visit


    def showMember(self):
        print(f'{self.memberID}, {self.first_name}, {self.surname}, {self.contact_number}, {self.last_visit}')

    def isActive(self):
        now = datetime.datetime.now()
        last_visit = datetime.datetime.strptime(self.last_visit, '%YYYY-%MM-%DD')

        if (now - last_visit).days < 30:
            return True
        else:
            return False
        
class NormalMember(Member):
    def __init__(self, memberID, first_name, surname, contact_number, memberType):
        super().__init__(memberID, first_name, surname, contact_number, memberType)

        self.stored_value = 0.00

    def get_stored_value(self):
        return self.stored_value
    
    def set_stored_value(self, value):
        self.stored_value = value

    def showMember(self):
        print(f'Normal: {self.memberID}, {self.first_name}, {self.surname}, {self.contact_number}, {self.last_visit}')

class AnnualMember(Member):
    def __init__(self, memberID, first_name, surname, contact_number, memberType):
        super().__init__(memberID, first_name, surname, contact_number, memberType)

        self.date_register = str(datetime.datetime.now())
        self.fee = 500

    def get_expiry_date(self):
        return self.expiry_date

    def set_expiry_date(self, expiry_date):
        self.expiry_date = expiry_date

    def showMember(self):
        print(f'Annual: {self.memberID}, {self.first_name}, {self.surname}, {self.contact_number}, {self.last_visit}')
```

```python
with open('members.txt') as f:
    data = []
    annual = []
    normal = []

    for line in f:
        line = line.strip().split(',')

        if line[4] == 'annual':
            annual += [AnnualMember(line[0], line[1], line[2], line[3], line[4])]
        else:
            normal += [NormalMember(line[0], line[1], line[2], line[3], line[4])]

for member in annual:
    member.showMember()

for member in normal:
    member.showMember()
```

```python
import sqlite3

query = "CREATE TABLE IF NOT EXISTS 'Member' \
    ('MemberID' TEXT PRIMARY KEY, \
    'FirstName' TEXT,'Surname' TEXT, \
    'ContactNo' TEXT, 'LastVisit' TEXT, \
    'MemberType' TEXT);"

with sqlite3.connect('JPgym.db') as conn:
    cursor = conn.cursor()

    cursor.execute('BEGIN TRANSACTION')
    cursor.execute(query)

    insert_annual = 'INSERT INTO Member VALUES (?, ?, ?, ?, ?, ?)'

    for member in annual:
        cursor.execute(insert_annual, (member.get_memberID(),
                                       member.get_first_name(),
                                       member.get_surname(),
                                       member.get_contact_number(),
                                       member.get_last_visit(),
                                       member.memberType))
        
    for member in normal:
        cursor.execute(insert_annual, (member.get_memberID(),
                                       member.get_first_name(),
                                       member.get_surname(),
                                       member.get_contact_number(),
                                       member.get_last_visit(),
                                       member.memberType))
        
    conn.commit()
```

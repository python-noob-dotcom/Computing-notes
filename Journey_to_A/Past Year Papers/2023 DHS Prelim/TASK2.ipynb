```python
class Node:
    def __init__(self):
        self.data = None
        self.next = None

    def get_data(self):
        return self.data
    
    def get_next(self):
        return self.next
    
    def set_data(self, val):
        self.data = val
    
    def set_next(self, nextNode):
        self.next = nextNode

class linkedlist:
    def __init__(self):
        self.head = None

    def add_to_end(self, data):
        new_node = Node()
        new_node.set_data(data)
        if self.head == None:
            self.head = new_node
            return

        else:
            curr = self.head

            while curr.get_next() != None:
                curr = curr.get_next()

            curr.set_next(new_node)
            return
    
    def search(self, target):
        if self.head == None:
            return False
        
        else:
            curr = self.head

            while curr != None:
                if curr.get_data() == target:
                    return True
                
                curr = curr.get_next()

            return False
        
    def get_values(self):
        if self.head == None:
            return 'Empty Linked List'
        
        curr = self.head
        data = []

        while curr != None:
            data += [curr.get_data()]
            curr = curr.get_next()

        return data
    
class HashTable:
    def __init__(self, size= 10):
        self.size = size
        self.array = [linkedlist() for i in range(self.size)]

    def hash(self, value):
        return value % self.size
    
    def insert(self, value):
        ll = self.array[self.hash(value)]
        ll.add_to_end(value)

    def search(self, target):
        return self.array[self.hash(target)].search(target)
    
    def display(self):
        for i in range(len(self.array)):
            print(f'index {i}: {self.array[i].get_values()}')

ht = HashTable(10)
for num in lst:
    ht.insert(num)

ht.display()
```

```python
def task2_3():
    global lst
    global ht

    for num in lst:
        ht.search(num)

task2_3()
```

```python
def merge_sort(lst):
    if len(lst) > 1:
        mid = len(lst) // 2

        left = lst[:mid]
        right = lst[mid:]

        merge_sort(left)
        merge_sort(right)

        i = j = k = 0

        while i < len(left) and j < len(right):
            if left[i] < right[j]:
                lst[k] = left[i]
                i += 1
            else:
                lst[k] = right[j]
                j += 1
            k += 1

        while i < len(left):
            lst[k] = left[i]
            i += 1
            k += 1

        while j < len(right):
            lst[k] = right[j]
            j += 1
            k += 1

    return lst

def binary_search(arr, target, low, high):
    if low < high:
        mid = (low + high) // 2

        if arr[mid] == target:
            return True

        if arr[mid] > target:
            return binary_search(arr, target, low, mid - 1)
        else:
            return binary_search(arr, target, mid + 1, high)
        
    return False

def task2_4():
    global lst

    lst_2 = lst
    lst_2 = merge_sort(lst_2)

    for num in lst:
        binary_search(lst_2, num, 0, len(lst_2) - 1)
```

```python
# import the built-in library timeit 
import timeit 

# use the timeit function to call the task2_3() procedure & run it just once 
time2_3 = timeit.timeit(lambda: task2_3(), number=1)
time2_4 = timeit.timeit(lambda: task2_4(), number=1)

# print out the time, in seconds
print(time2_3)
print(time2_4)
```

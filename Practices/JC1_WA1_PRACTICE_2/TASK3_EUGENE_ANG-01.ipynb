```python
# Task 3.1
# Generate the grid

def row(lst):
    return lst
def make_grid(val1, val2, val3, val4):
    lst = []
    lst.append(val1)
    lst.append(val2)
    lst.append(val3)
    lst.append(val4)
    return lst
# Output the grid in a "table format"
def display_row(grid_row):
    row_display = ''
    for value in grid_row:
        row_display += str(value)
        row_display += " "
    return row_display

def display_grid(grid):
    for row in grid:
        print(display_row(row))
```

```python
# Task 3.2
import random



# define functions of all the possible transformations

def transformation_1(grid):
    side = random.randint(1, 2)
    if side == 1:
            grid[0], grid[1] = grid[1], grid[0]
    if side == 2:
            grid[2], grid[3] = grid[3], grid[2]
    return grid

def transformation_2(grid):
    side = random.randint(1, 2)
    if side == 1:
            for row in grid:
                row[2], row[3] = row[3], row[2]
    if side == 2:
            for row in grid:
                row[0], row[1] = row[1], row[0]
    return grid

def transformation_3(grid):
    grid_2[0], grid_2[2] = grid_2[2], grid_2[0]
    grid_2[1], grid_2[3] = grid_2[3], grid_2[1]
    return grid

def transformation_4(grid):
    for row in grid_2:
            row[0], row[1], row[2], row[3] = row[3], row[2], row[1], row[0]
    return grid


# Main code

# init rows
row1 = row([4, 3, 2, 1])
row2 = row([1, 2, 4, 3])
row3 = row([3, 4, 1, 2])
row4 = row([2, 1, 3, 4])


grid = make_grid(row1, row2, row3, row4)

# init list of all possible transformations
transformations = ['swap_row', 'swap_column', 'swap_top_bottom', 'swap_left_right']
for i in range(10):
    grid_2 = grid
    transformation = random.choice(transformations)
    # Tuple assignment to swap rows
    if transformation == 'swap_row':
        grid_2 = transformation_1(grid_2)
    
    # Tuple assignment to swap columns
    elif transformation == 'swap_column':
        grid_2 = transformation_2(grid_2)

    # Tuple assignment to swap top quadrant rows and bottom quadrant rows
    elif transformation == 'swap_top_bottom':
        grid_2 = transformation_3(grid_2)
    
    # Tuple assignent to swap left columns and right columns
    elif transformation == 'swap_left_right':
        grid_2 = transformation_4(grid_2)

    print('-'*16)

    # Output new
    display_grid(grid_2)
```

```python
# Task 3.3
import random

grid = [[4, 3, 2, 1],
        [1, 2, 4, 3],
        [3, 4, 1, 2],
        [2, 1, 3, 4]]

transformations = ['swap_row', 'swap_column', 'swap_top_bottom', 'swap_left_right']
for i in range(100):
    grid_2 = grid
    transformation = random.choice(transformations)
    side = random.randint(1, 2)
    if transformation == 'swap_row':
        if side == 1:
            grid_2[0], grid_2[1] = grid_2[1], grid_2[0]
            commentary = 'Swaps two rows in the same quadrants'
        if side == 2:
            grid_2[2], grid_2[3] = grid_2[3], grid_2[2]
            commentary = 'Swap two rows in the same quadrant'
    elif transformation == 'swap_column':
        if side == 1:
            for row in grid_2:
                row[2], row[3] = row[3], row[2]
                
        if side == 2:
            for row in grid_2:
                row[0], row[1] = row[1], row[0]
        commentary = 'Swap two columns in the same quadrants'    
    elif transformation == 'swap_top_bottom':
        grid_2[0], grid_2[2] = grid_2[2], grid_2[0]
        grid_2[1], grid_2[3] = grid_2[3], grid_2[1]
        commentary = 'Swap top and bottom quadrant rows'
    elif transformation == 'swap_left_right':
        for row in grid_2:
            row[0], row[1], row[2], row[3] = row[3], row[2], row[1], row[0]
        commentary = 'Swap left and right quadrant rows'

    print('-'*16)
    for line in grid_2:
        print(f'{line[0]:<5}{line[1]:<5}{line[2]:<5}{line[3]:<5}')
    print(f'Transformation {transformations.index(transformation) + 1}: {commentary}')
```

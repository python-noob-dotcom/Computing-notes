```python
y, x = input('Enter the height and width of the field: ').split(',')

x, y = int(x), int(y)

mines = input('Enter the number of mines: ')

while int(mines) >= x*y:
    mines = input('Enter the number of mines thats is less than the number of squares in the grid: ')

grid = [['.' for i in range(x)] for i in range(y)]
display_grid = grid

import random

locations = []

while len(locations) < int(mines):
    coords = [random.randint(0, y - 1), random.randint(0, x - 1)]

    if coords not in locations:
        locations += [coords]

    else:
        pass

for location in locations:
    grid[location[0]][location[1]] = 'X'
```

```python
game_end = False
squares = len(grid) * len(grid[0]) - mines
played = 0

while not game_end:
    for row in display_grid:
        for char in row:
            print(char, end= '')
        print()
    
    x, y = input('Enter a row and column number to try: ').split(',')

    x, y = int(x), int(y)

    if grid[y][x] == 'X':
        game_end = True
        print('You Landed on the mine. Oh No!')

    else:
        played += 1
        counter = 0
        for i in range(y - 1, y + 2):
            if i < len(grid):
                for j in range(x - 1, x + 2):
                    if j < len(grid[0]):
                        if grid[y][x] == 'X':
                            counter += 1
                        else:
                            pass
                    else:
                        break
            else:
                break
        
        display_grid[y][x] = str(counter)

        if played == squares:
            for row in display_grid:
                for char in row:
                    print(char, end= '')
            
            print('You Win!')
            break
```

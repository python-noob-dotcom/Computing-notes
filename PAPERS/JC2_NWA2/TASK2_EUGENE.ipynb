```python
# Task 2.1

class Node:
    def __init__(self, key, value):
        self.key = key
        self.value = value
        self.nxt = None

class linkedlist:
    def __init__(self):
        self.head = None

    def insert(self, key, data):
        """
        Inserts key, value pair into the front of the linked list
        """
        new_node = Node(key, data)
        if self.head == None:
            self.head = new_node
            return 1
        
        else:
            new_node.nxt = self.head
            self.head = new_node
            return 1
    
    def search(self, key):
        """
        Searches for the key from the key value pairs in the linked list, returns None if not found, else returns the value as well as the 
        position of the linked list the key value pair is in
        """
        if self.head == None:
            return None
        
        current = self.head
        counta = 0

        while current != None:
            if current.key == key:
                return f'{current.value} found at position {counta}'
            
            counta += 1
            current = current.nxt
            
        return None
    
    def display(self):
        """
        Displays the key, value pairs in the linked list, returns [] if empty
        """
        disp = []
        current = self.head

        # Iterate through
        while current != None:
            disp.append((current.key, current.value))
            current = current.nxt

        return disp
```

```python
# Task 2.2

class HashTable:
    def __init__(self, size):
        self.size = size
        self.table = [linkedlist() for i in range(size)]

    def hash_algorithm(self, key):
        """
        Returns the hash value for the key
        """
        return key % self.size
    
    def hash_insert(self, key, value):
        """
        Implements chaining to insert key, value pair into the hash table
        """
        i = self.hash_algorithm(key)

        self.table[i].insert(key, value)

    def hash_search(self, key):
        """
        Implements searching for the key in the hash table
        """
        i = self.hash_algorithm(key)

        return self.table[i].search(key)
    
    def hash_display(self):

        """
        Displays the hashtable
        """
        for ll in self.table:
            print(ll.display())
```

```python
# Task 2.3
import csv

with open('./HASH.csv', 'r') as f:
    reader = csv.reader(f)

    data = []

    for line in reader:
        data.append(line)


hash_table = HashTable(25)

for line in data:
    hash_table.hash_insert(int(line[0]), line[1])

print(hash_table.hash_search(34))
print(hash_table.hash_search(59))
hash_table.hash_display()
```

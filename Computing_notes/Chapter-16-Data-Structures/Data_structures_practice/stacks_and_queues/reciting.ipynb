linked list

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.nxt = None

# Will not include the auxillary functions, except for is_empty
class linkedlist:
    def __init__(self):
        self.head = None

    def is_empty(self):
        return self.head == None

    def insert_front(self, data):
        new_node = Node(data)

        if self.is_empty():
            self.head = new_node

        else:
            new_node.nxt = self.head.nxt

            self.head = new_node

    def insert_back(self, data):
        new_node = Node(data)

        if self.is_empty():
            self.head = new_node

        else:
            current = self.head

            while current.nxt != None:
                current = current.nxt

            current.nxt = new_node

    def delete_front(self):

        if self.is_empty():
            return -1
        
        else:
            self.head = self.head.nxt

    def delete_back(self):

        if self.is_empty():
            return -1
        
        else:
            current = self.head

            while current.nxt.nxt != None:
                current = current.nxt

            current.nxt = None

    def insert_in_order(self, data):

        if self.is_empty():
            self.head = Node(data)

        else:

            current = self.head

            while current.data < data and current.nxt != None:
                current = current.nxt

            current.nxt = Node(data)

    def delete(self, data):

        if self.is_empty():
            return -1
        
        else:
            
            current = self.head

            while current.data != data:
                if current.nxt == None:
                    return -1

                previous = current
                current = current.nxt
            
            previous.nxt = None            
```

stacks

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.nxt = None

class stack:
    def __init__(self):
        self.top = None

    def is_empty(self):
        return self.top == None
    
    def push(self, data):
        new_node = Node(data)

        if self.is_empty():
            self.top = new_node

        else:
            new_node.nxt = self.top
            self.top = new_node

    def pop(self):

        if self.is_empty():
            return -1
        
        self.top = self.top.nxt

    def peek(self):
        if self.top != None:
            return self.top.data
        
        return 'Empty stack'
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.nxt = None

class queue:
    def __init__(self):
        self.front = None

    def is_empty(self):
        return self.front == None

    def enqueue(self, data):
        new_node = Node(data)

        if self.is_empty():
            self.front = new_node

        else:
            current = self.front

            while current.nxt != None:
                current = current.nxt

            current.nxt = new_node

    def dequeue(self):

        if self.is_empty():
            return -1
        
        a = self.front

        self.front = self.front.nxt

        return a
    
    def peek(self):
        return self.front.data if self.front != None else None
    
    def display(self):
        current = self.front

        while current != None:
            if current.nxt != None:
                print(current.data, end = ' --> ')
            else:
                print(current.data)
            current = current.nxt

class priorityNode:
    def __init__(self, data, priority):
        self.data = data
        self.priority = priority
        self.nxt = None

class priorityQueue:
    def __init__(self):
        self.front = None

    def is_empty(self):
        return self.front == None
    
    def enqueue(self, data, priority):
        new_node = priorityNode(data, priority)

        if self.is_empty():
            self.front = new_node

        elif self.front.priority > priority:
            new_node.nxt = self.front.nxt
            self.front = new_node

        else:
            current = self.front

            while current.priority < priority:
                if current.nxt == None:
                    break

                previous = current
                current = current.nxt
    
            new_node.nxt = previous.nxt

            previous.nxt = new_node

    def dequeue(self):
            
        if self.is_empty():
            return -1
        
        a = self.front

        self.front = self.front.nxt

        return a
    

class circularQueue:
    def __init__(self, capacity):
        self.front = -1
        self.rear = -1
        self.capacity = capacity
        self.queue = [None] * capacity

    def is_empty(self):
        return self.front == -1 and self.rear == -1
    
    def is_full(self):
        return (self.rear + 1) % self.capacity == self.front
    
    def enqueue(self, data):
        if self.is_full():
            return -1
        
        if self.is_empty():
            self.front = (self.front + 1 ) % self.capacity

        self.rear = (self.rear + 1) % self.capacity
        self.queue[self.rear] = data

    def dequeue(self):

        if self.is_empty():
            return -1
        
        a = self.queue[self.front]
        self.queue[self.front] = None

        self.front = (self.front + 1) % self.capacity

        if self.front == self.rear:
            self.front = -1
            self.rear = -1

        return a
    
        
```

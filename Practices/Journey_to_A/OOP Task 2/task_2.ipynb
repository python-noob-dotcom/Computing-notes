```python
class Time:
    def __init__(self, time_str):
        self.__total_sec = time_str

    def get_minute(self):
        return int
    
    def get_second(self):
        return int
    
    def display_time(self):
        return self.__total_sec
    
    def compare_time(self, other):
        return self.__total_sec > other.display_time()

    
```

```python
class BandingList:
    def __init__(self, test_item, op, boundary_values):
        self.__test_item = test_item
        self.__op = op
        self.__boundary_values = boundary_values

    def test_band(self):
        string = f'{self.__test_item + ":":<20}'

        for boundary in self.__boundary_values:
            string += str(boundary) + ' ' + '|'
        return string
    
    def calc_point(self, record):
        i = 0
        record = str(record)
        if self.__op == '<':
            bound = str(self.__boundary_values[i])
            while record > bound:
                i = i + 1
                if i + 1 > len(self.__boundary_values):
                    return 0
                else:
                    bound = str(self.__boundary_values[i])
        elif self.__op == '>':
            bound = str(self.__boundary_values[i])
            while record < bound:
                i = i + 1
                if i + 1 > len(self.__boundary_values):
                    return 0
                else:
                    bound = str(self.__boundary_values[i])

        return 5 - i

def trial_banding_list():
    b = BandingList('Sit ups', '>', [42, 37, 33, 28, 24])
    b.test_band()

b = BandingList('Sit ups', '>', [42, 37, 33, 28, 24])
print(b.calc_point(43)) # extreme bound
print(b.calc_point(23)) # extreme bound
print(b.calc_point(33)) # in the middle
```

```python
class TestStandard:
    def __init__(self, age, gender, bd_lists):
        self.age = age
        self.gender = gender
        self.bd_lists = []

        for bd_list in bd_lists:
            
            self.bd_lists += [BandingList(bd_list[0], bd_list[1], bd_list[2:])]

    def test_std(self):
        string = f'Age: {self.age}, Gender: {self.gender}\n'
        for bd_list in self.bd_lists:
            string += bd_list.test_band() + '\n'
        
        return string
    
def init_standard(file_name):
    data = []
    with open(file_name, 'r') as f:
        for line in f:
            data += [line.strip().split(',')]

    headers = data[0]
    operators = data[1]
    raw_data = data[2:]

    standards = []

    if 'female' in file_name:
        while len(raw_data) > 5:
            bd_lists = []
            for j in range(3, len(headers)):
                bd_list = [headers[j], operators[j]]
                for i in range(len(raw_data[:5])):
                    bd_list += [raw_data[i][j]]
                bd_lists += [bd_list]
                
                    
            standards += [TestStandard(raw_data[0][0], 'female', bd_lists)]
            raw_data = raw_data[5:]

    if 'male' in file_name:
        while len(raw_data) > 5:
            bd_lists = []
            for j in range(3, len(headers)):
                bd_list = [headers[j], operators[j]]
                for i in range(len(raw_data[:5])):
                    bd_list += [raw_data[i][j]]

                bd_lists += [bd_list]
                
                    
            standards += [TestStandard(raw_data[0][0], 'male', bd_lists)]
            raw_data = raw_data[5:]

    return standards

def trial_init_standard():
    b = init_standard('./test_standard_male.txt')

    for teststd in b:
        print(teststd.test_std())

trial_init_standard()
```

```python
class StudentRecord:
    def __init__(self, name, age, gender, test_results):
        self.name = name
        self.age = age
        self.gender = gender
        self.file = './test_standard_male.txt' if gender == 'M' else './test_standard_female.txt'
        self.test_results = test_results

    def process_score(self):
        scores = []
        score = 0
        band = ''

        standards = init_standard(self.file)
        for standard in standards:
            print(type(standard.age))
            if standard.age == self.age:
                standard_2 = standard

        for i in range(len(self.test_results)):
            scores += [standard_2.bd_lists[i].calc_point(self.test_results[i])]
            score += standard_2.bd_lists[i].calc_point(self.test_results[i])

        
        if min(scores) >= 3:
            if score >= 21:
                band = 'Gold'
            else:
                band = 'Silver'
        elif min(scores) <= 2:
            if min(scores) == 2:
                if score >= 15:
                    band = 'Silver'
                else:
                    band = 'Bronze'
            if min(scores) == 1:
                band = 'Bronze'
            else:
                band = 'No Award'

        return scores, band
    
    def results(self):
        scores, band = self.process_score()

        string = f'{self.name}, {self.gender}, {self.age}, {band}\n'

        for test_result in self.test_results:
            string += str(test_result) + '\t|'
        string += '\n'
        for score in scores:
            string += str(score) + '\t|'
    
        return string
    

def process_student_record():
    with open('./student_test_records.txt', 'r') as f:
        data = []
        i = 0
        for line in f:
            if i == 0:
                i += 1
                continue
            data += [line.strip().split(',')]
            i += 1

    female = []
    count_f_gold = 0
    count_f_bronze = 0
    count_f_silver = 0
    count_f_fail = 0
    male = []
    count_m_gold = 0
    count_m_bronze = 0
    count_m_silver = 0
    count_m_fail = 0

    for record in data:
        if record[1] == 'f':
            temp = StudentRecord(record[0], record[2], record[1], record[3:])
            female += [temp]
            if temp.process_score()[1] == 'Gold':
                count_f_gold += 1
            elif temp.process_score()[1] == 'Silver':
                count_f_silver += 1
            elif temp.process_score()[1] == 'Bronze':
                count_f_bronze += 1
            else:
                count_f_fail += 1
        else:
            temp = StudentRecord(record[0], record[2], record[1], record[3:])
            male += [temp]
            if temp.process_score()[1] == 'Gold':
                count_m_gold += 1
            elif temp.process_score()[1] == 'Silver':
                count_m_silver += 1
            elif temp.process_score()[1] == 'Bronze':
                count_m_bronze += 1
            else:
                count_m_fail += 1
        

    with open('./student_result.txt', 'w') as f:
        for record in female:
            f.write(record.results())
            f.write('\n')
        for record in male:
            f.write(record.results())
            f.write('\n')

        f.write(f'{"":<10}|{"No Award":<10}|{"Bronze":<10}|{"Silver":<10}|{"Gold":<10}\n{"Male":<10}|{count_m_fail:<10}|{count_m_bronze:<10}|{count_m_silver:<10}|{count_m_gold:<10}\n{"Female":<10}|{count_f_fail:<10}|{count_f_bronze:<10}|{count_f_silver:<10}|{count_f_gold:<10}')

process_student_record() 
```

```python
# Task 2.1
import random

def createIsland(rowsize, colsize):
    island = [['.' for _ in range(colsize)] for _ in range(rowsize)]
    island[0][0] = '*'
    return island

def plantCoconuts(grid):
    col_size = len(grid)
    row_size = len(grid[0])

    coconuts = []
    while len(coconuts) < 3:
        coords = [(random.randint(0, col_size - 1), random.randint(0, row_size - 1))]
        if coords not in coconuts:
            coconuts += coords
            grid[coords[0][0]][coords[0][1]] = 'C'
    return grid

def display(grid):
    for row in grid:
        for i in range(len(row)):
            marker = row[i]
            if i != len(row) - 1:
                print(marker, end = ' ')
            else:
                print(marker)
            

def move(grid, current, direction, distance, coconut):
    current = list(current)
    direction = direction.upper()
    distance = int(distance)
    current_copy = [current[i] for i in range(len(current))]
    if direction == 'L':
            current[1] += -1 * distance
    if direction == 'R':
            current[1] += 1 * distance
    if direction == 'U':
            current[0] += 1 * distance
    if direction == 'D':
            current[0] += -1 * distance

    # if move up
    if current[0] > current_copy[0]:
        for i in range(0, current[0] - current_copy[0], -1):
            if grid[current_copy[0] + i][current_copy[1]] == 'C':
                coconut += 1
                grid[current_copy[0] + i][current_copy[1]] = 'E'
                continue
            grid[current_copy[0] + i][current_copy[1]] = '*'

    # if move down
    if current[0] < current_copy[0]:
        for i in range(current_copy[0] - current[0]) :
            if grid[current_copy[0] + i][current_copy[1]] == 'C':
                coconut += 1
                grid[current_copy[0] + i][current_copy[1]] = 'E'
                continue
            grid[current_copy[0] + i][current_copy[1]] = '*'

    # if move right
    if current[1] > current_copy[1]:
        for i in range(current[1] - current_copy[1]):
            if grid[current_copy[0]][current_copy[1] + i] == 'C':
                coconut += 1
                grid[current_copy[0]][current_copy[1] + i] = 'E'
                continue
            grid[current_copy[0]][current_copy[1] + i] = '*'

    # if move left
    if current[1] < current_copy[1]:
        for i in range(0, current[1] - current_copy[1], -1):
            if grid[current_copy[0]][current_copy[1] + i] == 'C':
                coconut += 1
                grid[current_copy[0]][current_copy[1] + i] = 'E'
                continue
            grid[current_copy[0]][current_copy[1] + i] = '*'
    
    return (grid, tuple(current), coconut)
```

```python
# Task 2.2
import os

def main():
    grid = createIsland(10, 15)
    grid = plantCoconuts(grid)
    current = (0, 0)
    moves = 0
    coconut = 0
    display(grid)

    while moves < 8:
        movement = input('Enter the move u make with the format <direction> <space>: ').split(' ')
        grid, current, coconut = move(grid, current, movement[0], movement[1], coconut)
        os.system('cls')
        display(grid)

    if coconut < 3 or current != (9, 14):
        print('You have won!')
        print('You have reach finish point and eat 3 coconut')

    else:
        print('Sorry u never win')
        if coconut < 3:
            print('You never eat coconut')
        if current != (9, 14):
            print('You never reach finish')

if __name__ == '__main__':
    main()
```

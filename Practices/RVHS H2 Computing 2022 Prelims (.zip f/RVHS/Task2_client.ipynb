```python
# Task 2.1
def validate_guesscode(guesscode):
    if not len(guesscode) == 4:
        return False
    if guesscode.isdigit():
        for char in guesscode:
            if char not in "123456":
                return False
        return True
    return False

# Client version 1 with delay(1)
# For Task 2.2
import socket, time

def menu():
    game_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    print("Client initiates connection to server.")
    game_socket.connect(('127.0.0.1', 12345))
    print("Client connected to server.")

    done = False
    while not done:
        print("Master Mind Game Menu")
        print("1) Just start a new game")
        print("2) Make a guess code")
        print("3) Display history")
        print("4) Quit")

        option = input("Type an option:")
        while len(option)!=1 and option not in "1234":
            option = input("Please type an valid option:")
        if option == "1":
            tx_msg = "START\n"
            game_socket.send(tx_msg.encode())

            rx_msg_str = ""
            while "\n" not in rx_msg_str:
                rx_msg_str += game_socket.recv(1024).decode()
            message = rx_msg_str[:-1]

            header, attempts = message.split(",")
            print("New game starts begins with " + attempts + " attempts.")

        elif option =="2":
            guesscode = input("Make a guess:")
            while not validate_guesscode(guesscode):
                guesscode = input("Make an valid guess (Your guess must be of 4 digits and each digits must be of range 1 - 6):")
            tx_msg = "GUESS," + guesscode + "\n"
            game_socket.send(tx_msg.encode())

            rx_msg_str = ""
            while "\n" not in rx_msg_str:
                rx_msg_str += game_socket.recv(1024).decode()
            message = rx_msg_str[:-1]

            message_lst = message.split(",")
            if message_lst[0] == "GUESS":
                print("Guess Result: " + message_lst[1])
            elif message_lst[0] == "WON":
                done = True
                print("You have WON. The code is " + guesscode + ".")
            else:
                done = True
                print("You have run out of attemps. The code is " + message_lst[1] + ".")
        elif option =="3":
            tx_msg = "HISTORY\n"
            game_socket.send(tx_msg.encode())

            rx_msg_str = ""
            while "\n" not in rx_msg_str:
                rx_msg_str += game_socket.recv(1024).decode()
            message = rx_msg_str[:-1]

            message_lst = message.split(",")
            print("You have " + message_lst[1] + " attempts left.")
            count = 1
            for i in range(2, len(message_lst)-1, 2):
                print(count, message_lst[i], message_lst[i+1])
                count += 1

        elif option =="4":
            tx_msg = "QUIT\n"
            game_socket.send(tx_msg.encode())
            done = True
            print("Game program ends.")
        else:
            print("Invalid option entered.")

    game_socket.close()
menu()
```
